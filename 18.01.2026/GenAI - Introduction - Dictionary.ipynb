# Data Structures - Dictionaries

Another useful data type built into Python is the dictionary (see Mapping Types — dict). Dictionaries are sometimes found in other languages as “associative memories” or “associative arrays”. Unlike sequences, which are indexed by a range of numbers, dictionaries are indexed by keys, which can be any immutable type; strings and numbers can always be keys. Tuples can be used as keys if they contain only strings, numbers, or tuples; if a tuple contains any mutable object either directly or indirectly, it cannot be used as a key. You can’t use lists as keys, since lists can be modified in place using index assignments, slice assignments, or methods like append() and extend().

It is best to think of a dictionary as a set of key: value pairs, with the requirement that the keys are unique (within one dictionary). A pair of braces creates an empty dictionary: {}. Placing a comma-separated list of key:value pairs within the braces adds initial key:value pairs to the dictionary; this is also the way dictionaries are written on output.

Source: https://docs.python.org/3/tutorial/datastructures.html

```python
# Basic dictionary
person = {"name": "Alice", "age": 25, "city": "New York"}

print(person)
print(person["name"])
print(person["age"])
print(person["city"])
```

```python
# Nested dictionary
student = {
    "name": "Kumar",
    "grades": {"math": 90, "science": 85, "history": 88}
}

print(student)
print(student['grades'])
print(student['grades']['math'])
```

```python
# Dictionary with list values
inventory = {
    "fruits": ["apple", "banana", "orange"],
    "vegetables": ["carrot", "broccoli"]
}

print(inventory)
print(inventory['fruits'])
print(inventory['vegetables'])
```

```python
# Dictionary comprehension
squares = {x: x**2 for x in range(5)}
print(squares)
```

```python
# Dictionary with mixed types
info = {"name": "Bob", "is_active": True, "balance": 100.5}
print(info)
print(info['name'])
print(info['is_active'])
print(info['balance'])
```

```python
# Iterating over key-value pairs
for key, value in person.items():
    print(f"{key}: {value}")
    
#Any guess why we use f? 
# A formatted string literal or f-string is a string literal that is prefixed with 'f' or 'F'. 
# These strings may contain replacement fields, which are expressions delimited by curly braces {}.
# While other string literals always have a constant value, 
# formatted strings are really expressions evaluated at run time.
```

```python
# Using fromkeys()
keys = ["a", "b", "c"]
default_dict = dict.fromkeys(keys, 1)
print(default_dict)
default_dict1 = dict.fromkeys(keys, 22)
print(default_dict1)
```

```python
# Merging dictionaries
dict1 = {"a": 1, "b": 2}
dict2 = {"c": 3, "d": 4}
merged = {*dict1, *dict2}
merged = {**dict1, **dict2}


print('dict1',dict1)
print('dict2',dict2)
print('merged',merged)
```

```python
# Dictionary length
print(len(merged))
```

```python
# Dictionary with integer keys
numbers = {1: "one", 2: "two", 3: "three"}
print(numbers)
```

```python
# Dictionary with tuple keys (immutable)
coordinates = {(0, 0): "Origin", (1, 2): "Point A"}
print(coordinates)
print(coordinates[(0,0)])
print(coordinates[(1,2)])
```

```python
# Dictionary with different data types as values
profile = {"name": "Eve", "age": 30, "hobbies": ["reading", "hiking"], "married": True}
print(profile)
```

```python
# Using dict() constructor
user = dict(username="admin", password="1234")
print(user)
```

```python
# Dictionary with default values
defaults = dict.fromkeys(["a", "b", "c"], "default")
print(defaults)
print(defaults['a'])
print(defaults['b'])
print(defaults['c'])
```

```python
#Counting word frequency
sentence = "hello world hello python"
word_count = {}
for word in sentence.split():
    word_count[word] = word_count.get(word, 0) + 1
    print('word_count[word]', word_count[word])
    print('word_count',word_count)
    
print(word_count)
```

```python
#Dictionary with functions as values
operations = {
    "add": lambda x, y: x + y,
    "multiply": lambda x, y: x * y,
    "divide": lambda x, y: y / x,    
    "power": lambda x, y: y**x     
}
print(operations["add"](3, 4))
print(operations["multiply"](3, 4))
print(operations["divide"](3, 4))
print(operations["power"](3, 4))
```

```python
# Accessing keys and values
car = {"brand": "Toyota", "model": "Corolla", "year": 2020}
print(car.keys())
print(car.values())
```

```python
# Dictionary update()
car.update({"color": "blue", "year": 2021})
print('car',car)
```

```python
#Using setdefault()
settings = {}
settings.setdefault("theme", "dark")
print(settings)
```

```python
# when key exists
person = {"name": "Alice", "age": 25}

# 'name' exists, so it returns the existing value
name = person.setdefault("name", "Unknown")
print(name)           
print(person)   
```

```python
#when key does not exist
person = {"name": "Alice", "age": 25}

# 'city' doesn't exist, so it adds 'city' with the default value 'New York'
city = person.setdefault("city", "New York")
print(city)           
print(person)        
```

```python
 #Dictionary sorted by keys
scores = {"Alice": 85, "job": 90, "Charlie": 80, "Elephant": 100, "Logistics": 122, 'Apple':1, "zebra":45, "Truck":90}
sorted_scores = dict(sorted(scores.items()))
print(sorted_scores)
```

```python
# Dictionary sorted by values
sorted_by_values = dict(sorted(scores.items(), key=lambda item: item[1])) # try changin this value to 0 and see
print('sorted_by_values',sorted_by_values)
```

```python
#list of dictionaries
classroom = [
    {"name": "Anna", "grade": "A"},
    {"name": "Ben", "grade": "B"}
]

print(classroom)
print(classroom[0])
print(classroom[1])
```

```python
# Extracting names way 1
names = []
for student in classroom:
    names.append(student["name"])

print(names)
```

```python
# Extracting names way 2
names = [student["name"] for student in classroom]
print(names) 

names_1 = [student["grade"] for student in classroom]
print(names_1) 
```

```python
# we can use Map also
names = list(map(lambda student: student["name"], classroom))
print(names)


names_2 = list(map(lambda student: student["grade"], classroom))
print(names_2)
```

```python
A = (list(set(student['major'] for student in students_info.values()))) #list of all student majors without duplicates
print(A)
# Set keeps only one unique occurrence of each.
```

```python
B = (max(students_info, key=lambda x: students_info[x]['gpa'])) #student with the highest GPA.
print(B)
```
