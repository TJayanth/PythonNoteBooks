# Pandas Framework

```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
}

df = pd.DataFrame(data)
print("DataFrame:")
print(df)
```

```python
print("\nNames:")
print(df['Name'])# Accessing columns
```

```python
print("\nFirst row:")
print(df.loc[0])# Accessing rows by index
```

```python
df['Salary'] = [70000, 80000, 90000]
print("\nDataFrame with Salary:")
print(df)# Adding a new column
```

```python
high_salary = df[df['Salary'] > 75000] # Selecting
print("\nPeople with high salary:")
print(high_salary)
```

```python
print("\nAverage age:", df['Age'].mean())# some statistics
```

```python
sorted_df = df.sort_values('Age')
print("\nDataFrame sorted by Age:")
print(sorted_df) # Sorting DataFrame by Age
```

```python
name_city = df[['Name', 'City']]# Selecting specific columns
print("\nName and City columns:")
print(name_city) 
```

```python
city_group = df.groupby('City')['Salary'].mean() # Grouping data and calculating mean
print("\nAverage salary by city:")
print(city_group) 
```

```python
data_with_nan = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, None, 35],
    'City': ['New York', 'Los Angeles', 'Chicago'],
    'Salary': [70000, 80000, None]
}

df_nan = pd.DataFrame(data_with_nan) # Handling missing data
print("\nDataFrame with missing values:")
print(df_nan)
```

```python
df_filled = df_nan.fillna({'Age': df_nan['Age'].mean(), 'Salary': df_nan['Salary'].mean()}) # Filling missing values
print("\nDataFrame with missing values filled:")
print(df_filled) # replace the missing values with mean
```

```python

df_dropped = df_nan.dropna()
print("\nDataFrame with missing values dropped:")
print(df_dropped)# removed rows with missing values
```

```python
import pandas as pd
# pip install pandas

# Creating a Pandas Series from a list
data = [1, 2, 3, 4, 5]
series = pd.Series(data)
print('series',series)

# Creating a Pandas Series with a specified index
index = ['a', 'b', 'c', 'd', 'e']
series_with_index = pd.Series(data, index=index)
print('series_with_index',series_with_index)

# Creating a Pandas Series from a dictionary
data_dict = {'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5}
series_from_dict = pd.Series(data_dict)
print('series_from_dict',series_from_dict)

# Accessing data in a Series
print(series[2])  # Accessing element at index 2
print(series_with_index['b'])  # Accessing element with index 'b'
print(series_from_dict['b'])  # Accessing element with index 'b'
```

```python
print('series',series)

# Return the first n rows
first_n_rows = series.head(3)
print('first_n_rows',first_n_rows)

# Return the last n rows
last_n_rows = series.tail(3)
print('last_n_rows',last_n_rows)

# Return dimensions (Rows, columns)
dimensions = series.shape
print('dimensions',dimensions)

# Generate descriptive statistics
stats = series.describe()
print('stats',stats)

# Return unique values
unique_values = series.unique()
print('unique_values',unique_values)

# Return the number of unique values
num_unique_values = series.nunique()
print('num_unique_values',num_unique_values)
```

```python
# Let us consider another simple series and see some more functionalities in pandas

series = pd.Series([1, 2, 3, 4, 5, 6, None, 8, 9, 10,10,8,6,3,2,4])

# Count non-NA/null values
count_non_null = series.count()
print('\ncount_non_null:', count_non_null)
```

```python
total_sum = series.sum()
print('\ntotal_sum:', total_sum)

mean_value = series.mean()
print('\nmean_value:', mean_value)

median_value = series.median()
print('\nmedian_value:', median_value)

std_dev = series.std()
print('\nstd_dev:', std_dev)

min_value = series.min()
print('\nmin_value:', min_value)

max_value = series.max()
print('\nmax_value:', max_value)

idx_max = series.idxmax()
print('\nidx_max:', idx_max)

idx_min = series.idxmin()
print('\nidx_min:', idx_min)

value_counts = series.value_counts()
print('\nvalue_counts:\n', value_counts)

null_check = series.isnull()
print('\nnull_check:\n', null_check)

non_null_check = series.notnull()
print('\nnon_null_check:\n', non_null_check)

filled_series = series.fillna(0)  # Replace with 0, or any value
print('\nfilled_series:\n', filled_series)

dropped_nulls = series.dropna()
print('\ndropped_nulls:\n', dropped_nulls)


squared_series = series.apply(lambda x: x**2 if pd.notnull(x) else x) # square each element except none/null
print('\nsquared_series:\n', squared_series)

cum_sum = series.cumsum() # Cumulative sum
print('\ncum_sum:\n', cum_sum)

cum_prod = series.cumprod()
print('\ncum_prod:\n', cum_prod)

sorted_series = series.sort_values()
print('\nsorted_series:\n', sorted_series)

sorted_index = series.sort_index()
print('\nsorted_index:\n', sorted_index)

clipped_series = series.clip(lower=2, upper=8)
print('\nclipped_series:\n', clipped_series)
```

```python
# Querying a series in pandas

import pandas as pd

data = {'a': 10, 'b': 20, 'c': 30, 'd': 40, 'e': 50}
series = pd.Series(data)


selected_greater_than_30 = series[series > 30]# Select elements greater than 30


selected_equal_to_20 = series[series == 20]# Select elements equal to 20


selected_not_equal_to_40 = series[series != 40]# Select elements not equal to 40


selected_multiple_conditions = series[(series > 20) & (series < 50)]# Select elements based on multiple conditions


selected_by_list = series[series.isin([20, 40, 60])]# Select elements based on a list of values


string_series = pd.Series(['apple', 'banana', 'cherry', 'date', 'elderberry'])
selected_by_string_method = string_series[string_series.str.startswith('b')]# Select elements using string methods (if applicable)


selected_by_index_labels = series.loc[['a', 'c', 'e']]# Query based on index labels


selected_by_numeric_position = series.iloc[1:4]# Query based on numeric position


print("Original Series:")
print(series)
print("\nSelected greater than 30:")
print(selected_greater_than_30)
print("\nSelected equal To 20:")
print(selected_equal_to_20)
print("\nSelected not equal to 40:")
print(selected_not_equal_to_40)
print("\nSelected based on multiple conditions:")
print(selected_multiple_conditions)
print("\nSelected based on list of values:")
print(selected_by_list)
print("\nSelected based on string method (startswith):")
print(selected_by_string_method)
print("\nSelected based on index labels:")
print(selected_by_index_labels)
print("\nSelected based on numeric position:")
print(selected_by_numeric_position)
```

```python
import pandas as pd

# Step 1: Load the housing dataset
df = pd.read_csv('Housing_Data.csv')

# Step 2: Overview of the DataFrame
print("First few rows of the dataset:")
print(df.head())

print("\nDataFrame Info:")
print(df.info())

print("\nSummary Statistics of Numerical Columns:")
print(df.describe())

# Step 3: Calculate descriptive statistics for selected columns
numerical_columns = ['price', 'lotsize', 'bedrooms', 'bathrms', 'stories', 'garagepl']

# Mean, Median, Standard Deviation
mean_values = df[numerical_columns].mean()
print("\nMean values for numerical columns:")
print(mean_values)

median_values = df[numerical_columns].median()
print("\nMedian values for numerical columns:")
print(median_values)

std_values = df[numerical_columns].std()
print("\nStandard deviation for numerical columns:")
print(std_values)

# Step 4: Determine correlation with price
correlations = {
    'lotsize_price': df['lotsize'].corr(df['price']),
    'bedrooms_price': df['bedrooms'].corr(df['price']),
    'bathrms_price': df['bathrms'].corr(df['price']),
    'stories_price': df['stories'].corr(df['price']),
    'garagepl_price': df['garagepl'].corr(df['price'])
}
print("\nCorrelations with Price:")
print(correlations)

# Step 5: Count occurrences in categorical columns
categorical_columns = ['driveway', 'recroom', 'fullbase', 'gashw', 'airco', 'prefarea']

for column in categorical_columns:
    print(f"\nValue counts for {column}:")
    print(df[column].value_counts())
```

```python
import pandas as pd

# 1. Create a Pandas Series for sales data
# - Use a list of daily sales figures to create a Pandas Series
# - Assign days of the week as the index

sales_data = [120, 150, 130, 170, 160, 180, 140]
days_of_week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Create the Series
sales_series = pd.Series(data=sales_data, index=days_of_week)
print("Sales Series:\n", sales_series)

# 2. Access and manipulate sales data

# - Access sales data for specific days using index labels
wednesday_sales = sales_series['Wednesday']
print("\nSales on Wednesday:", wednesday_sales)

# - Calculate total sales for the week
total_sales = sales_series.sum()
print("\nTotal sales for the week:", total_sales)

# - Identify the day with the highest and lowest sales
highest_sales_day = sales_series.idxmax()
lowest_sales_day = sales_series.idxmin()
print("\nDay with the highest sales:", highest_sales_day)
print("Day with the lowest sales:", lowest_sales_day)

# 3. Basic analysis of sales data

# - Calculate the average sales for the week
average_sales = sales_series.mean()
print("\nAverage sales for the week:", average_sales)

# - Determine any days with sales figures significantly different from the average

significantly_larger_sales = sales_series[(sales_series > average_sales)]
significantly_smaller_sales = sales_series[(sales_series < average_sales)]
print("\nDays with significantly_larger_sales figures:\n", significantly_larger_sales)
print("\nDays with significantly_smaller_sales figures:\n", significantly_smaller_sales)

# Let's consider "significantly different" as sales more than 20% above or below the average
threshold =  0.2*average_sales
significantly_different_sales = sales_series[(sales_series > average_sales + threshold) | (sales_series < average_sales - threshold)]
print("\nDays with significantly different sales figures:\n", significantly_different_sales)
```
