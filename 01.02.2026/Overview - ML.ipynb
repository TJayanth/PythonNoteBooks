# Regression – House Price Prediction


```python

import pandas as pd
from sklearn.linear_model import LinearRegression

# Simple dataset
data = {
    "Size_sqft": [800, 1000, 1200, 1500, 1800],
    "Price": [40, 50, 65, 80, 95]
}

df = pd.DataFrame(data)

X = df[["Size_sqft"]]
y = df["Price"]

model = LinearRegression()
model.fit(X, y)

# Predict new house price
predicted_price = model.predict([[1400]])
print("Predicted price for 1400 sqft house:", predicted_price[0])
```

### The model learned a relationship between size and price, then predicted a new value 

# Classification – Spam Detection

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer

texts = [
    "Win money now",
    "Limited offer click here",
    "Meeting at 10 AM",
    "Project discussion tomorrow"
]

labels = [1, 1, 0, 0]  # 1 = Spam, 0 = Not Spam

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(texts)

model = MultinomialNB()
model.fit(X, labels)

test_email = ["Win a free prize now"]
test_vec = vectorizer.transform(test_email)

prediction = model.predict(test_vec)
print("Spam" if prediction[0] == 1 else "Not Spam")
```

### “We converted text into numbers, then trained a classifier

# Unsupervised Learning – Customer Clustering

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Customer data: Spending vs Visits
data = {
    "Spending": [200, 220, 250, 600, 650, 700],
    "Visits": [2, 3, 2, 8, 9, 10]
}

df = pd.DataFrame(data)

kmeans = KMeans(n_clusters=2, random_state=42)
df["Cluster"] = kmeans.fit_predict(df)

plt.scatter(df["Spending"], df["Visits"], c=df["Cluster"])
plt.xlabel("Spending")
plt.ylabel("Visits")
plt.title("Customer Segmentation")
plt.show()
```

### We never gave labels. The algorithm grouped similar customers automatically

# Recommendation System – Simple Movie Recommende

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# User-Movie ratings
ratings = np.array([
    [5, 4, 0, 1],
    [4, 0, 0, 1],
    [1, 1, 0, 5],
    [0, 1, 5, 4]
])

similarity = cosine_similarity(ratings)
print("User similarity matrix:\n", similarity)
```

### Users with similar preferences influence recommendations.

# NLP – Sentiment Analysis

```python
from sklearn.linear_model import LogisticRegression

texts = [
    "I love this product",
    "This is amazing",
    "Worst experience ever",
    "I hate it"
]

labels = [1, 1, 0, 0]  # 1 = Positive, 0 = Negative

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(texts)

model = LogisticRegression()
model.fit(X, labels)

test_text = ["I really love it"]
test_vec = vectorizer.transform(test_text)

sentiment = model.predict(test_vec)
print("Positive" if sentiment[0] == 1 else "Negative")
```

### “ChatGPT does this at a much larger scale.”

# Reinforcement Learning – Learning by Trial & Error

```python
import numpy as np

# Actions: 0 = Left, 1 = Right
Q = np.zeros(2)
learning_rate = 0.001
rewards = [0, 1]  # Right gives reward

# Training loop
for episode in range(100):
    action = np.random.choice([0, 1])
    Q[action] = Q[action] + learning_rate * (rewards[action] - Q[action])

print("Learned Q-values:", Q)
print("Best action:", "Right" if np.argmax(Q) == 1 else "Left")
```

```python
# Agent tries both actions
# Gets reward feedback
# Updates behavior
```

```python
### The agent tries actions, gets rewards, and slowly learns what works best.”
```

# ANN – Artificial Neural Network

```python
from sklearn.neural_network import MLPClassifier

# Study hours vs result
X = [[1], [2], [3], [4], [5], [6]]
y = [0, 0, 0, 1, 1, 1]  # 0 = Fail, 1 = Pass

model = MLPClassifier(
    hidden_layer_sizes=(5,),
    activation='relu',
    max_iter=2000,
    random_state=42
)

model.fit(X, y)

prediction = model.predict([[4.5]])
print("Prediction:", "Pass" if prediction[0] == 1 else "Fail")
```

```python
# Visualizing ANN Decision Boundary
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([[1],[2],[3],[4],[5],[6]])
y = np.array([0,0,0,1,1,1])

model = MLPClassifier(hidden_layer_sizes=(5,), max_iter=2000)
model.fit(X, y)

x_range = np.linspace(0, 7, 100).reshape(-1,1)
y_pred = model.predict(x_range)

plt.scatter(X, y, label="Actual")
plt.plot(x_range, y_pred, label="ANN Prediction")
plt.legend()
plt.title("ANN Learning Pattern")
plt.show()
```

# Same data → different answers

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
import numpy as np

X = [[1],[2],[3],[4],[5],[6]]
y = [0,0,0,1,1,1]

dt = DecisionTreeClassifier()
lr = LogisticRegression()

dt.fit(X, y)
lr.fit(X, y)

print("Decision Tree:", dt.predict([[3.5]]))
print("Logistic Regression:", lr.predict([[3.5]]))
```

```python
“Different models can give different answers on the same data.”
------> Model choice matters
```

# Memorizing vs Learning

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

X = np.array([1,2,3,4,5]).reshape(-1,1)
y = np.array([2,3,5,8,12])

poly = PolynomialFeatures(degree=4)
X_poly = poly.fit_transform(X)

model = LinearRegression()
model.fit(X_poly, y)

plt.scatter(X, y)
plt.plot(X, model.predict(X_poly))
plt.title("Overfitting Example")
plt.show()
```

# Data Matters More Than Algorithm
## Garbage data → garbage results

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Clean data
X = [[1],[2],[3],[4]]
y = [2,4,6,8]

model = LinearRegression()
model.fit(X, y)
print("Clean prediction:", model.predict([[5]]))

# Noisy data
y_noisy = [20, 1, 50, 3]
model.fit(X, y_noisy)
print("Noisy prediction:", model.predict([[5]]))
```

```python
### ML doesn’t understand truth. It understands patterns
```

# Train vs Test – Why we don’t trust training accuracy

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X = [[i] for i in range(1, 21)]
y = [0]*10 + [1]*10

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

model = LogisticRegression()
model.fit(X_train, y_train)

print("Train accuracy:", accuracy_score(y_train, model.predict(X_train)))
print("Test accuracy:", accuracy_score(y_test, model.predict(X_test)))
```
