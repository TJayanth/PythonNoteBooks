# Tensor

# A tensor is simply a multi-dimensional array used to represent data such as images, batches of images, or videos in machine learning.

| Structure | Order / Rank | Dimensions | Example | Representation | Typical Use |
|-----------|--------------|------------|---------|----------------|-------------|
| **Scalar** | Rank 0 | Single value | 5, 3.14 | $x$ | Temperature value, loss value |
| **Vector** | Rank 1 | 1D array | [2, 4, 6] | $x_i$ | Feature list, word embeddings |
| **Matrix** | Rank 2 | 2D array (rows × columns) | [[1,2],[3,4]] | $x_{ij}$ | Images, datasets, linear transformations |
| **3D Tensor** | Rank 3 | 3D array | Stack of matrices | $x_{ijk}$ | RGB images (height × width × channels) |
| **4D Tensor** | Rank 4 | 4D array | 4D array | $x_{ijkl}$ | CNN input (batch × height × width × channels) |
| **n-D Tensor** | Rank n | n-dimensional array | Higher dimensional data | $x_{i_1,i_2,...,i_n}$ | Deep learning models |

```python
# For your understanding 

| Concept | Example in AI                                          |
| ------- | ------------------------------------------------------ |
| Scalar  | Model loss value                                       |
| Vector  | Word embedding (e.g., 300 numbers representing a word) |
| Matrix  | Dataset with rows = samples and columns = features     |
| Tensor  | Video data (frames × height × width × channels)        |
```

```python
# Example 

| Data Type       | Tensor Shape                          |
| --------------- | ------------------------------------- |
| Grayscale Image | (Height, Width)                       |
| RGB Image       | (Height, Width, 3)                    |
| Batch of Images | (Batch Size, Height, Width, Channels) |
| Video           | (Frames, Height, Width, Channels)     |
```

```python
# Tensor Types

| Data Type       | Tensor Shape                          | Tensor Type   | Rank / Order  | Explanation                                         |
| --------------- | ------------------------------------- | ------------- | ------------- | --------------------------------------------------- |
| Grayscale Image | (Height, Width)                       | **Matrix**    | Rank 2 Tensor | Each pixel has one intensity value                  |
| RGB Image       | (Height, Width, 3)                    | **3D Tensor** | Rank 3 Tensor | 3 channels represent Red, Green, Blue               |
| Batch of Images | (Batch Size, Height, Width, Channels) | **4D Tensor** | Rank 4 Tensor | Multiple images processed together in deep learning |
| Video           | (Frames, Height, Width, Channels)     | **4D Tensor** | Rank 4 Tensor | Sequence of image frames forming a video            |
```

# Why Tensors Are Important

Tensors are fundamental in modern machine learning, deep learning, physics, and scientific computing because they allow us to represent and process multi-dimensional data efficiently.

```python
| Reason                                | Explanation                                                      | Example                                 |
| ------------------------------------- | ---------------------------------------------------------------- | --------------------------------------- |
| **Represent complex data**            | Real-world data is often multi-dimensional                       | Image = height × width × channels       |
| **Efficient computation**             | Hardware like GPUs and TPUs are optimized for tensor operations  | Deep learning training                  |
| **Generalization of data structures** | Tensors unify scalars, vectors, and matrices under one framework | Scalar → Vector → Matrix → Tensor       |
| **Parallel processing**               | Tensor operations can run in parallel on thousands of GPU cores  | Neural network training                 |
| **Core of deep learning frameworks**  | Almost all ML libraries operate on tensors                       | Neural network inputs, weights, outputs |
```

# Modern AI frameworks like TensorFlow and PyTorch internally represent all data as tensors.

How the Concept of Tensors Came Up

The concept did not originate in AI. It actually came from mathematics and physics.

# 1. Origin in Mathematics (19th Century)
Tensors emerged from linear algebra and differential geometry when mathematicians needed a way to represent quantities with multiple directions and dimensions.
One of the key contributors was
Gregorio Ricci-Curbastro, who developed tensor calculus in the late 1800s.

# 2. Use in Physics
Tensors became very important in physics, especially in describing physical systems.
For example, Albert Einstein used tensors extensively in the General Theory of Relativity to describe how gravity affects spacetime.
In physics:
•	Scalars → temperature
•	Vectors → velocity
•	Tensors → stress, curvature of spacetime

# 3. Entry into Computer Science and AI
With the rise of machine learning and neural networks, researchers needed structures that could represent:
•	Images (3D data)
•	Video (4D data)
•	Large datasets
•	Model parameters
Tensors were ideal because they naturally support multi-dimensional arrays and mathematical operations.


```python
# 1. Example: Vector → Tensor

This shows how a 1D vector becomes a tensor.
```

```python
import torch

# Creating a vector
vector = torch.tensor([2, 4, 6, 8])

print("Vector:")
print(vector)

print("\nTensor Shape:")
print(vector.shape)

# Shape [4]
# This is a Rank-1 tensor (Vector)
```

```python
# 2. Example: Image → Tensor

from PIL import Image
import numpy as np

# Load image
image = Image.open("TESTIMAGE.jpg")

# Convert image to tensor-like array
image_tensor = np.array(image)

print("Tensor Shape:")
print(image_tensor.shape)
```

```python
# 3. Example: Batch of Images → Tensor

Deep learning models process multiple images together.
```

```python
import torch
import torchvision.transforms as transforms
from PIL import Image
import os

folder = "dataset"

# Resize all images to same size
transform = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.ToTensor()
])

images = []

for file in os.listdir(folder):
    if file.endswith(".jpg") or file.endswith(".png"):
        img_path = os.path.join(folder, file)
        image = Image.open(img_path)

        tensor_img = transform(image)
        images.append(tensor_img)

batch_tensor = torch.stack(images)

print("Tensor Shape:", batch_tensor.shape)
```

```python
# 4. Visualizing the Image Tensor

import matplotlib.pyplot as plt

plt.imshow(image_tensor)
plt.title("Image represented as Tensor")
plt.show()
```
