# Linear Transformation

```python
A linear transformation is a function that takes a vector as input and produces another vector as output, while preserving the structure of space.

In linear algebra, matrices represent linear transformations.
Mathematically:
T(v)=Av

Where:
	A → transformation matrix
	v → input vector
	Av → transformed vector
```

```python

A = np.array([[2,0],
              [0,1]])

v = np.array([1,2])

result = A @ v

print(result)
```

```python
1. The Vector
v = [1, 2]
This means the point is located at:
x = 1
y = 2

2.  The Transformation Matrix
A = [[2,0],
     [0,1]]
This matrix performs scaling.
    
It means:
x → 2x
y → 1y
So:
	x direction doubles
	y direction stays the same
```

```python
import matplotlib.pyplot as plt
import numpy as np

A = np.array([[2,0],
              [0,1]])

v = np.array([1,2])

v_transformed = A @ v

plt.quiver(0,0,v[0],v[1],angles='xy',scale_units='xy',scale=1,label="Original")
plt.quiver(0,0,v_transformed[0],v_transformed[1],angles='xy',scale_units='xy',scale=1,label="Transformed")

plt.xlim(0,4)
plt.ylim(0,4)
plt.grid()
plt.show()
```

# Determinant Indicates Scaling
A matrix represents a linear transformation of space. When that transformation is applied to vectors, it changes the area (in 2D) or volume (in 3D) of shapes. The absolute value of the determinant tells how much the space is scaled.

Example
If det⁡(A)=3

then the transformation scales areas by 3.
	Unit square area = 1
	After transformation = 3
If det⁡(A)=0.5

the space shrinks to half the area.

| Determinant | Meaning                                   |
| ----------- | ----------------------------------------- |
| det = 3     | area expands 3×                           |
| det = 0.5   | area shrinks to half                      |
| det = -2    | area expands 2× **and flips orientation** |
| det = 0     | space collapses → not invertible          |


# Determinant Indicates Invertibility

A matrix is invertible if it has an inverse.
The rule is simple:
det⁡(A)≠0 ➡ matrix is invertible
det⁡(A)=0 ➡ matrix is NOT invertible

Why?
If determinant = 0, the transformation collapses space.
Examples:
	a square becomes a line
	a cube becomes a plane

    
Information is lost → cannot reverse the transformation

```python
# Matrix Inverse

A = np.array([[1,2],
              [3,4]])

A_inv = np.linalg.inv(A)

print(A_inv)
```

```python
Think of the matrix inverse as the operation that undoes a transformation. 
If a matrix A transforms vectors, then its inverse A⁻¹ reverses that transformation.
Mathematically:
A^(-1) A=I

and
AA^(-1)=I

Where I is the identity matrix.
```

```python
# To check whethet its actually inverse 
# A−1 A=I

print(A @ A_inv)  # Identify matrix
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Transformation matrix
A = np.array([[1,2],
              [3,4]])

# Original vector
v = np.array([1,2])

# Transform the vector
v_transformed = A @ v

# Compute inverse matrix
A_inv = np.linalg.inv(A)

# Recover original vector
v_recovered = A_inv @ v_transformed

# Create subplots
fig, axes = plt.subplots(1,3, figsize=(12,4))

# Original vector
axes[0].quiver(0,0,v[0],v[1],angles='xy',scale_units='xy',scale=1)
axes[0].set_xlim(-1,6)
axes[0].set_ylim(-1,6)
axes[0].grid()
axes[0].set_title("Original Vector")

# Transformed vector
axes[1].quiver(0,0,v_transformed[0],v_transformed[1],angles='xy',scale_units='xy',scale=1)
axes[1].set_xlim(-1,12)
axes[1].set_ylim(-1,12)
axes[1].grid()
axes[1].set_title("After Transformation (A @ v)")

# Recovered vector
axes[2].quiver(0,0,v_recovered[0],v_recovered[1],angles='xy',scale_units='xy',scale=1)
axes[2].set_xlim(-1,6)
axes[2].set_ylim(-1,6)
axes[2].grid()
axes[2].set_title("Recovered Vector (A⁻¹ @ v')")

plt.show()
```

# What is a Linear Equation?



A linear equation is an equation where the variables appear only to the first power (power = 1).

General form:
ax+b=0

or with two variables:

ax+by=c

Where:
	x, y → variables
	a, b, c → constants

The key idea:
No powers like x², x³
No products like xy

That is why it is called linear.

# Geometric Meaning

In 2D, a linear equation represents a straight line.
Example:

y=2x+1

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10,10,100)
y = 2*x + 1

plt.plot(x,y)
plt.grid()
plt.title("Linear Equation: y = 2x + 1")
plt.show()
```

```python
#Linear Equation with Multiple Variables

Example:
2x+y=5

This is still linear, but now it involves two variables.
    
To find a unique solution we need another equation:
x+3y=6

Now we have a system of linear equations.
    
# This is where linear algebra and matrices become powerful.

# so, why called a System ? Wondered why ? 
A system simply means a collection or group of equations.
    
Example:
    
2x+y=5
x+3y=6

These equations are linked together because we want values of xand y that satisfy both equations at the same time.
That is why it is called a system.
```

## Geometric Interpretation

Each equation represents a line in 2D.

Example:
Equation 1: 2x+y=5

Equation 2: x+3y=6

Each line has infinitely many solutions individually.
But the solution of the system is the point where the lines intersect.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5,5,100)

y1 = 5 - 2*x      # from 2x + y = 5
y2 = (6 - x)/3    # from x + 3y = 6

plt.plot(x,y1,label="2x + y = 5")
plt.plot(x,y2,label="x + 3y = 6")

plt.legend()
plt.grid()
plt.title("System of Linear Equations")

plt.show()
```

```python
Finally! why professors use this AX=B !!!!

When we have many equations:

2x+y=5
x+3y=6

we write them as:
Ax=b

which allows us to solve them using matrices and linear algebra.
```

# Solving Systems of Linear Equations

```python


This is actually one of the main reasons matrices exist.
Example system:
2x+y=5
x+3y=6

Instead of solving manually, we convert it to matrix form.
Ax=b

Where:
A=[([2 1] [1 3])]
x=[(x y)]
b=[(5 6)]
```

```python
import numpy as np

# Coefficient matrix
A = np.array([[2,1],
              [1,3]])

# Result vector
b = np.array([5,6])

# Solve Ax = b
x = np.linalg.solve(A,b)

print("Solution:", x)
```

```python
So

x = 1.8
y = 1.4
```

```python
#For verification

print(A @ x)
```

```python
The key idea:

Ax = b

Multiply both sides by A⁻¹:

A^(-1) Ax=A^(-1) b
Ix=A^(-1) b
x=A^(-1) b

So solving linear equations is basically:
                            x = A⁻¹ b
    
But in practice we use:

np.linalg.solve()

because it is numerically more stable.
```

```python
#Another example
A = np.array([[1,2,3],
              [2,1,1],
              [3,4,2]])

b = np.array([14,10,19])

x = np.linalg.solve(A,b)

print(x)
```

```python
print(A @ x)
```

```python
# Linear algebra allows us to solve many equations simultaneously using matrices
```

```python
so  Instead of doing:

x = np.linalg.inv(A) @ b    # this is what i had mentioned in the equation
You do:
x = np.linalg.solve(A, b)  #  correct way

Instead, it uses efficient methods like:

Gaussian elimination
LU decomposition


Why did i use the word “numerically stable” ?
Computers store numbers with limited precision (rounding errors happen).

Some operations:
    amplify tiny errors
    others keep them small
                                                
Using inverse:
np.linalg.inv(A) @ b

Problems:
    Computing inverse is expensive
    Introduces extra rounding errors                                      
    Errors get magnified
                                                
Using solve():
np.linalg.solve(A, b)

Benefits:
    Avoids computing inverse entirely
    Uses stable algorithms
    Gives more accurate results
                                                
Intuition
Think of it like this:

inv(A) @ b → "Let me completely reverse A, then multiply" (overkill)
solve(A, b) → "Let me directly find x efficiently" (smart way )

```

# Angle Between Two Vectors

```python

This concept connects three things you already know:
	Dot Product
	Vector Norm (length)
	Geometry

--------------------> # The angle between two vectors tells us how similar their directions are.<----------------------------------

# Mathematical Formula
    
The angle θ between vectors v and w is:

cos⁡(θ)=(v⋅w)/(∣∣v∣∣" " ∣∣w∣∣)

Where:
	v⋅w→ dot product
	∣∣v∣∣→ length of vector v
	∣∣w∣∣→ length of vector w

Then we compute:

θ=〖cos⁡〗^(-1) ((v⋅w)/(∣∣v∣∣" " ∣∣w∣∣))

```

```python
import numpy as np

v = np.array([3,4])
w = np.array([4,1])

dot_product = np.dot(v,w)

norm_v = np.linalg.norm(v)
norm_w = np.linalg.norm(w)

cos_theta = dot_product / (norm_v * norm_w)

theta = np.arccos(cos_theta)

print("Angle in radians:", theta)
print("Angle in degrees:", np.degrees(theta))
```

```python
import numpy as np
import matplotlib.pyplot as plt

v = np.array([3,4])
w = np.array([4,1])

plt.quiver(0,0,v[0],v[1],angles='xy',scale_units='xy',scale=1,label="v")
plt.quiver(0,0,w[0],w[1],angles='xy',scale_units='xy',scale=1,label="w")

plt.xlim(0,6)
plt.ylim(0,6)

plt.grid()
plt.legend()

plt.title("Angle Between Two Vectors")

plt.show()
```

# Method 2 — Determinant Test (Very Nice Trick !!!)

```python

For vectors in 2D:
v_1=[a,b]
v_2=[c,d]

Make a matrix:
A=[[a, b] [c,d]]

Compute determinant:
"det"(A)=ad-bc

Rule
det = 0 → vectors lie on same line (dependent)
det ≠ 0 → vectors are independent
```

```python
import numpy as np

v1 = np.array([1,2])
v2 = np.array([2,4])

A = np.column_stack((v1,v2))

print(np.linalg.det(A))
```

# Vector Projection
How much of one vector lies in the direction of another vector?

```python
Imagine a light shining on a vector, creating a shadow on another vector.

That shadow is the projection.

Example :
v  ------>
      \
       \
        w

The shadow of v on w is the projection of v onto w.
```

```python
Mathematical Formula

Projection of v onto w:

proj(v)=( (v⋅w)/(∣∣w∣∣^2 ) ) w

Where:
	v⋅w→ dot product
	∣∣w∣∣→ length of vector w
```

```python
import numpy as np

v = np.array([3,4])
w = np.array([4,1])

projection = (np.dot(v,w) / np.dot(w,w)) * w

print("Projection of v onto w:", projection)
```

```python
Example output:

[3.7588 0.9447]

This is the component of v in the direction of w.
```

```python
import numpy as np
import matplotlib.pyplot as plt

v = np.array([3,4])
w = np.array([4,1])

proj = (np.dot(v,w) / np.dot(w,w)) * w

plt.figure(figsize=(6,6))

plt.quiver(0,0,v[0],v[1],angles='xy',scale_units='xy',scale=1,color='blue',label="v")
plt.quiver(0,0,w[0],w[1],angles='xy',scale_units='xy',scale=1,color='green',label="w")
plt.quiver(0,0,proj[0],proj[1],angles='xy',scale_units='xy',scale=1,color='red',label="Projection of v on w")

plt.xlim(-1,6)
plt.ylim(-1,6)

plt.grid()
plt.legend()

plt.title("Vector Projection")

plt.show()
```

```python
You will see 

    vector v
    vector w
    projection of v on w
```

```python
# Just an added information
Projection splits vector v into two parts:
        v = (v-parallel) + (v-perpendicular)

Where:
	v-parallel → projection on w
	v-perpendicular → orthogonal component

# Thus, vector projection tells us how much of one vector lies in the direction of another vector.
```

```python
Projection is used in:

    least squares regression
    machine learning
    PCA
    computer graphics
    signal processing
```

```python
# jsut a clarification here 
import numpy as np

v = np.array([3,4])
w = np.array([4,1])

# projection (parallel component)
v_parallel = (np.dot(v,w) / np.dot(w,w)) * w

# orthogonal component
v_perpendicular = v - v_parallel

print("Projection:", v_parallel)
print("Orthogonal component:", v_perpendicular)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# vectors
v = np.array([4,3])
w = np.array([5,1])

# projection (parallel component)
v_parallel = (np.dot(v,w) / np.dot(w,w)) * w

# orthogonal component
v_perpendicular = v - v_parallel

plt.figure(figsize=(6,6))

# original vector
plt.quiver(0,0,v[0],v[1],angles='xy',scale_units='xy',scale=1,color='blue',label="v (original)")

# projection
plt.quiver(0,0,v_parallel[0],v_parallel[1],angles='xy',scale_units='xy',scale=1,color='green',label="Projection on w")

# orthogonal component
plt.quiver(v_parallel[0],v_parallel[1],v_perpendicular[0],v_perpendicular[1],
           angles='xy',scale_units='xy',scale=1,color='red',label="Orthogonal component")

# reference vector
plt.quiver(0,0,w[0],w[1],angles='xy',scale_units='xy',scale=1,color='purple',label="w")

plt.xlim(-1,7)
plt.ylim(-1,7)

plt.grid()
plt.legend()

plt.title("Vector Decomposition: v = v_parallel + v_perpendicular")

plt.show()
```

# Vector Span

This will help you understand how vectors generate space.

```python
# What is Span?

The span of vectors is the set of all vectors you can create using linear combinations.

For vectors v1 and v2:
        span(v1,v2)=c1.v1+c2.v2
where:
    c1,c2 ∈ R

------------------------------>>>>> # This means we can scale and add vectors to generate new ones.
```

```python
# Example

Let

v1=[1,0]
v2=[0,1]

Then
c1 [1,0]+c2[0,1]=[c1,c2]

So these two vectors can generate every point in the 2D plane.
    
They span the plane.
```

```python
import numpy as np
import matplotlib.pyplot as plt

v1 = np.array([1,0])
v2 = np.array([0,1])

points = []

for a in np.linspace(-2,2,10):
    for b in np.linspace(-2,2,10):
        p = a*v1 + b*v2
        points.append(p)

points = np.array(points)

plt.figure(figsize=(6,6))

plt.scatter(points[:,0], points[:,1], alpha=0.5)

plt.quiver(0,0,v1[0],v1[1],angles='xy',scale_units='xy',scale=1,color='red',label="v1")
plt.quiver(0,0,v2[0],v2[1],angles='xy',scale_units='xy',scale=1,color='green',label="v2")

plt.xlim(-3,3)
plt.ylim(-3,3)

plt.grid()
plt.legend()

plt.title("Span of Two Vectors")

plt.show()

# We can generate any combination of vectors !
```

```python
Important Cases of Span

Case 1 — Dependent vectors
Example:
v_1=[1,2]
v_2=[2,4]

Span → a line only
Because they lie on the same direction.

Case 2 — Independent vectors
Example:
v_1=[1,0]
v_2=[0,1]

Span → entire 2D plane
```

```python
# Span tells us what space a set of vectors can generate.
```

```python
Why Span Matters

Span explains:
    feature spaces
    dimensionality
    basis
    PCA
    embedding spaces
```

# Basis


```python
A basis tells us the minimum set of vectors needed to generate a space.
    
So if span answers:
What space can these vectors generate?
    
Basis answers:
What is the smallest set of vectors needed to generate that space?
```

```python
A set of vectors is a basis if:

    1 The vectors are linearly independent
    2 They span the space

Both conditions must be satisfied.
```

```python
Example: Basis for 2D Space

The most common basis is:

v1=[1,0]
v2=[0,1]

These are called the standard basis vectors.
    
Any vector in 2D can be written as:
    [x,y]=x[1,0]+y[0,1]

Example:
    [3,4]=3[1,0]+4[0,1]
```

```python
import numpy as np

v1 = np.array([1,0])
v2 = np.array([0,1])

# represent a vector using the basis
v = 3*v1 + 4*v2

print(v)
```

# So the basis vectors can construct any vector in the plane.

```python
import numpy as np
import matplotlib.pyplot as plt

v1 = np.array([1,0])
v2 = np.array([0,1])

plt.figure(figsize=(6,6))

plt.quiver(0,0,v1[0],v1[1],angles='xy',scale_units='xy',scale=1,color='red',label="Basis vector 1")
plt.quiver(0,0,v2[0],v2[1],angles='xy',scale_units='xy',scale=1,color='green',label="Basis vector 2")

# example vector constructed from basis
v = 3*v1 + 2*v2

plt.quiver(0,0,v[0],v[1],angles='xy',scale_units='xy',scale=1,color='blue',label="3v1 + 2v2")

plt.xlim(-1,5)
plt.ylim(-1,5)

plt.grid()
plt.legend()

plt.title("Basis Vectors Generating the Plane")

plt.show()
```

```python
| Space      | Basis Size |
| ---------- | ---------- |
| Line       | 1 vector   |
| Plane (2D) | 2 vectors  |
| 3D space   | 3 vectors  |

------------------------>  # The number of basis vectors = dimension of the space. <---------------------------
```

```python
# Example of NOT a Basis

v1 = [1,2]
v2 = [2,4]

These are dependent, so they cannot form a basis.

They only span a line, not the plane.
```

# A basis is the smallest set of independent vectors that can generate a space

```python
import numpy as np
import matplotlib.pyplot as plt

# basis vectors
v1 = np.array([1,0])
v2 = np.array([0,1])

points = []

# generate linear combinations
for a in np.linspace(-3,3,15):
    for b in np.linspace(-3,3,15):
        p = a*v1 + b*v2
        points.append(p)

points = np.array(points)

plt.figure(figsize=(6,6))

# generated points
plt.scatter(points[:,0], points[:,1], alpha=0.5)

# basis vectors
plt.quiver(0,0,v1[0],v1[1],angles='xy',scale_units='xy',scale=1,color='red',label="Basis v1")
plt.quiver(0,0,v2[0],v2[1],angles='xy',scale_units='xy',scale=1,color='green',label="Basis v2")

plt.xlim(-4,4)
plt.ylim(-4,4)

plt.grid()
plt.legend()

plt.title("Basis Vectors Spanning the Plane")

plt.show()
```

```python
Each point is created as:

a.v1+b.v2

for different values of aand b.

This visually proves:

    span(v1,v2)=R^2


By scaling and combining just two vectors, we can generate every point in the plane.
That is why they form a basis of the 2D space.
```

# Dimension of a Vector Space

```python
The dimension of a vector space tells us:
    How many independent directions exist in that space.
```

```python
Mathematically:
    Dimension=Number of vectors in a basis
```

```python
1 Line (1D)

Example vector:
v=[1]

Only one direction exists.
Basis:
{[1]}

Dimension: 1
```

```python
Plane (2D)

Example basis:

v_1=[1,0]
v_2=[0,1]

These two vectors generate the entire plane.

Dimension: 2

```

```python
3D Space

Basis:
v_1=[1,0,0]
v_2=[0,1,0]
v_3=[0,0,1]

Dimension:
3

```

```python
# Dimension = number of independent directions.

| Space    | Independent directions | Dimension |
| -------- | ---------------------- | --------- |
| Line     | 1                      | 1         |
| Plane    | 2                      | 2         |
| 3D space | 3                      | 3         |
```

```python
import numpy as np

v1 = np.array([1,0])
v2 = np.array([0,1])

A = np.column_stack((v1,v2))

print("Matrix:\n", A)
print("Rank:", np.linalg.matrix_rank(A))
```

```python
import numpy as np
import matplotlib.pyplot as plt

v1 = np.array([1,0])
v2 = np.array([0,1])

plt.figure(figsize=(6,6))

plt.quiver(0,0,v1[0],v1[1],angles='xy',scale_units='xy',scale=1,color='red',label="v1")
plt.quiver(0,0,v2[0],v2[1],angles='xy',scale_units='xy',scale=1,color='green',label="v2")

plt.xlim(-2,2)
plt.ylim(-2,2)

plt.grid()
plt.legend()

plt.title("Two Independent Directions → Dimension = 2")

plt.show()
```

```python
Rank = how many independent columns (or rows) a matrix has

OR

Rank = dimension of the space spanned by its columns

| Concept   | Meaning              |
| --------- | -------------------- |
| Dimension | Property of a space  |
| Rank      | Property of a matrix |

Rank = dimension of the column space of the matrix

So:

Matrix creates a space (via its columns)
Rank tells you the dimension of that space
```
