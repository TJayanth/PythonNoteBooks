```python
### **Line Plot**
# Line plots are used to show trends over time or continuous data.
# Typically used for time series data, stock market analysis, or sensor data visualization.

import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline 

time = np.linspace(0, 10, 100)
signal = np.sin(time)

plt.figure(figsize=(8, 5))
plt.plot(time, signal, label='Sine Wave', color='blue')
plt.title('Line Plot Example')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()

print(time)
```

```python
### **Bar Plot**
# Bar plots are used to represent categorical data with rectangular bars.
# Best for comparing values across categories like sales, revenue, or survey results.

import seaborn as sns

categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 8, 5, 9]

plt.figure(figsize=(8, 5))
sns.barplot(x=categories, y=values, palette='viridis')
plt.title('Bar Plot Example')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

```python
### **Histogram**
# Histograms are used to visualize the distribution of a continuous variable.
# Useful for understanding data distribution like income levels or exam scores.

data = np.random.randn(10)

plt.figure(figsize=(8, 5))
plt.hist(data, bins=10, color='purple', edgecolor='black')
plt.title('Histogram Example')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
print(data)
```

```python
### **Scatter Plot**
# Scatter plots are used to show the relationship between two variables.
# Often used in regression analysis or correlation studies.

x = np.random.rand(100)
y = np.random.rand(100)

plt.figure(figsize=(8, 5))
plt.scatter(x, y, color='green', alpha=0.6)
plt.title('Scatter Plot Example')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.grid(True)
plt.show()
```

```python
### **Box Plot**
# Box plots display the spread and skewness of the data using five summary statistics.
# Used for detecting outliers and comparing distributions.

# A Box Plot (also called a Whisker Plot) is a visual representation of the distribution of a dataset. 
# It displays key summary statistics such as:
#     Minimum
#     First Quartile (Q1)
#     Median (Q2)
#     Third Quartile (Q3)
#     Maximum
#     Outliers (using individual points)



import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Adjust the path to where the file is located
file_path = 'Iris.csv'

# Load data
iris = pd.read_csv(file_path)

# Box Plot for Sepal Length across species
plt.figure(figsize=(8, 5))
sns.boxplot(x='Species', y='SepalLengthCm', data=iris)
plt.title('Box Plot of Sepal Length across Species')
plt.show()

# Box Plot for Petal Length
plt.figure(figsize=(8, 5))
sns.boxplot(x='Species', y='PetalLengthCm', data=iris)
plt.title('Box Plot of Petal Length across Species')
plt.show()

# interpretation 
#     Box: Represents the interquartile range (IQR = Q3 - Q1), which contains the middle 50% of the data.
#     Line in the Box: The median (Q2) of the data.
#     Whiskers: Extend to the smallest and largest values within 1.5x the IQR.
#     Dots Outside the Whiskers: Outliers, representing values that lie beyond 1.5x IQR from the quartiles.
#     Width of the Box: Indicates the spread or variability.

# Example:
#     If the median is closer to the bottom of the box, the data is positively skewed.
#     If the median is closer to the top, it is negatively skewed.
#     Outliers are represented as dots or small circles.


# When Not to Use a Box Plot
#     If the dataset is too small: A box plot may not reveal patterns accurately.
#     When data has a high number of categories: It becomes cluttered.
#     For visualizing precise values: Use bar plots or scatter plots instea
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
file_path = 'Iris.csv'
iris = pd.read_csv(file_path)

# Plot the Box Plot
plt.figure(figsize=(8, 6))
ax = sns.boxplot(x='Species', y='SepalLengthCm', data=iris, palette='Set2')

# Add annotations for key components
plt.annotate('Median (Q2)', xy=(0, 5.1), xytext=(0.2, 5.5),
             arrowprops=dict(facecolor='blue', arrowstyle='->'), fontsize=10, color='blue')

plt.annotate('Upper Quartile (Q3)', xy=(1, 6.3), xytext=(1.2, 6.8),
             arrowprops=dict(facecolor='green', arrowstyle='->'), fontsize=10, color='green')

plt.annotate('Lower Quartile (Q1)', xy=(2, 5.7), xytext=(2.2, 5.2),
             arrowprops=dict(facecolor='orange', arrowstyle='->'), fontsize=10, color='orange')

plt.annotate('Whisker (Max)', xy=(0, 7.9), xytext=(0.2, 7.5),
             arrowprops=dict(facecolor='purple', arrowstyle='->'), fontsize=10, color='purple')

plt.annotate('Whisker (Min)', xy=(1, 4.5), xytext=(1.2, 4.8),
             arrowprops=dict(facecolor='purple', arrowstyle='->'), fontsize=10, color='purple')

plt.annotate('Outliers', xy=(2, 7.5), xytext=(2.2, 7.2),
             arrowprops=dict(facecolor='red', arrowstyle='->'), fontsize=10, color='red')

# Customize
plt.title('Box Plot of Sepal Length by Species with Annotations')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.show()

# Median (Q2)
#     The line inside the box represents the median (the middle value of the dataset).
#     It is annotated in blue.
# Upper Quartile (Q3)
#     The top of the box indicates the third quartile (Q3), representing the 75th percentile.
#     Annotated in green.

# Lower Quartile (Q1)
#     The bottom of the box is the first quartile (Q1), representing the 25th percentile.
#     Annotated in orange.

# Whiskers (Min & Max)
#     Whiskers extend to the smallest and largest data points within 1.5x IQR (Interquartile Range).
#     Shown in purple.

# Outliers
#     Points beyond the whiskers are classified as outliers and plotted individually.
#     They are marked in red.

```

```python

### Pie Chart
# Pie charts are used to show proportions and percentages.
# Best used when comparing parts of a whole, like market share or budget distribution.

# To represent categorical data as proportions.
# To compare market share, survey results, or budget allocations.
# When you have 5 or fewer categories. (Too many slices make it difficult to interpret.)
# When understanding the relative size of each part is importan

sizes = [30, 20, 25, 25]
labels = ['Product A', 'Product B', 'Product C', 'Product D']
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']

plt.figure(figsize=(8, 5))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.title('Pie Chart Example')
plt.show()
```

```python
### **Heatmap**
# Heatmaps visualize the correlation between variables using colors.
# Useful for correlation matrices and visualizing tabular data.

# A Heatmap is a data visualization technique that uses color to represent the values of a matrix. It’s particularly useful for visualizing:
#     Correlation between features
#     Missing values in a dataset
#     Magnitude of data values in a matrix

# What we can use it for ?
# Correlation Analysis: To observe how strongly two numerical variables are related.
# Feature Selection: Identify highly correlated features that might be redundant.
# Anomaly Detection: Spot extreme values using color variations.
# Trend Analysis: Track variations across time or categories.

# Why Correlation?
# Correlation measures the relationship between two variables.
#     Range: -1 to +1
#         +1: Strong positive correlation
#         -1: Strong negative correlation
#         0: No correlation



import numpy as np

# Calculate correlation matrix
correlation_matrix = iris.drop(columns=['Id', 'Species']).corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlation Matrix')
plt.show()


# Dark Blue (Negative Correlation):
#     When one variable increases, the other decreases.
#     Example: If PetalLengthCm and SepalWidthCm show a negative correlation, larger petals might correspond to narrower sepals.
# Dark Red (Positive Correlation):
#     When one variable increases, the other also increases.
#     Example: PetalLengthCm and PetalWidthCm usually have a high positive correlation.
# Light Colors (Near 0 Correlation):
#      No significant relationship between the variables.
```

```python
# A Pair Plot is a grid of scatter plots that shows pairwise relationships between numerical variables in a dataset.
# It is particularly useful for exploratory data analysis (EDA) to identify patterns, correlations, and class separability.
# In a Pair Plot:
#     Each cell in the grid represents a scatter plot between two different variables.
#     The diagonal cells typically display the distribution of each variable using histograms or KDE (Kernel Density Estimation) plots.
#     It often uses color (hue) to differentiate between categories, like species in the Iris dataset.

# A pair plot consists of:
#     Diagonal Plots
#     Off-Diagonal Plots

# A. Diagonal Plots (diag_kind='kde')
#     The diagonal shows the KDE plot (like a smooth histogram) for each numeric feature.
#     It helps us understand the distribution of data for each feature across different species.
#     Example:
#         A narrow peak indicates that most data points are close to a specific value (low variance).
#         A wide peak suggests more spread-out data (high variance).

# B. Off-Diagonal Plots (Scatter Plots)
#     Each cell shows the scatter plot between two variables.

#     It helps identify:
#         Linear Relationships: Variables that are positively or negatively correlated.
#         Class Separation: Some pairs of variables may clearly separate the species into distinct clusters.
#         Outliers: Points that deviate significantly from the cluster.

#     Example Interpretation:
#         PetalLengthCm vs PetalWidthCm usually shows excellent class separation for Iris species.
#         SepalLengthCm vs SepalWidthCm may have overlapping classes, especially between Iris-versicolor and Iris-virginica.


# what we can get

# Class Separation: Identify which variables best separate the species.
# Feature Relationships: Spot correlations or dependencies.
# Anomalies: Detect outliers and inconsistencies.
# Multicollinearity: Identify variables that are strongly correlated, which might be redundant.


import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the data
file_path = 'Iris.csv'
iris = pd.read_csv(file_path)

# Select specific columns (2 to 5) + Species
selected_columns = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm', 'Species']
iris_selected = iris[selected_columns]

# Plot the Pair Plot
plt.figure(figsize=(10, 10))
sns.pairplot(iris_selected, hue='Species', diag_kind='kde', markers=['o', 's', 'D'])

# Display the plot
plt.suptitle('Pair Plot for Selected Features of Iris Dataset', y=1.02)
plt.show()
```

```python
### **Area Plot**
# Area plots are similar to line plots but with the area below the line filled.
# Used to visualize cumulative data, like cumulative sales or energy consumption.

plt.figure(figsize=(10, 6))

# Plot Sepal and Petal lengths as area plots
plt.fill_between(iris['Id'], iris['SepalLengthCm'], color='blue', alpha=0.5, label='Sepal Length')
plt.fill_between(iris['Id'], iris['PetalLengthCm'], color='orange', alpha=0.5, label='Petal Length')

plt.title('Area Plot of Sepal and Petal Length')
plt.xlabel('Sample ID')
plt.ylabel('Length (cm)')
plt.legend()
plt.show()
```
