```python
## 2. Class Method
    Uses @classmethod

class Student:
    school = "IIT Madras"

    @classmethod
    def get_school(cls):
        return cls.school

# Works with the class itself, not individual objects
# Uses cls instead of self
```

```python
# Another example
class Bank:
    bank_name = "ABC Bank"   # class variable

    @classmethod
    def change_bank_name(cls, new_name):
        cls.bank_name = new_name   # modifies class variable
```

```python
Bank.change_bank_name("XYZ Bank")

print(Bank.bank_name)   # XYZ Bank
```

```python
# Modify class-level data
# Create alternative constructors
# Work with variables shared across all objects
```

```python
## 3. Static Method
No access to class or instance
```

```python
class Student:
    @staticmethod
    def is_adult(age):
        return age >= 18
        

# Independent method (no access to class or object)
# Key idea: No self, no cls
```

```python
# Another example 
class BankUtils:

    @staticmethod
    def validate_amount(amount):
        return amount > 0
```

```python
print(BankUtils.validate_amount(100))   # True
```

```python
# Utility/helper functions
# Logic not dependent on class or instance
```

```python
| Feature       | Instance Method | Class Method      | Static Method  |
| ------------- | --------------- | ----------------- | -------------- |
| First param   | `self`          | `cls`             | None           |
| Access object | ✅ Yes           | ❌ No              | ❌ No      |
| Access class  | ✅ Yes           | ✅ Yes             | ❌ No      |
| Use case      | Object logic    | Class-level logic | Utility/helper |
```

##  FYI :  @-decorator
(# we had already looked at this)

```python
A decorator is a function that modifies another function (without changing its code directly).

Simple idea: “Wrap a function and add extra behavior”

Why the @ ?
the @ is just syntactic sugar (shortcut) for applying a decorator.


Think of a decorator like:

Wrapping a gift ........

Gift = function
Wrapper = decorator
You didn’t change the gift, but you enhanced it
```

```python
# Without @ (actual working)

def my_decorator(func):
    def wrapper():
        print("Before function runs")
        func()
        print("After function runs")
    return wrapper

def say_hello():
    print("Hello!")

say_hello = my_decorator(say_hello)

say_hello()
```

```python
# With @ (clean way)

@my_decorator
def say_hello():
    print("Hello!")

say_hello()
```

```python
class BankAccount:
    
    def __init__(self, name, balance=0):
        # Constructor method: runs automatically when object is created
        self.name = name
        # Store account holder's name in the object
        self.balance = balance
        # Store initial balance (default is 0 if not given)

    def deposit(self, amount):
        # Method to add money to the account
        if amount > 0:
            # Check if deposit amount is valid (positive)
            self.balance += amount
            # Increase balance by deposit amount
            print(f"{amount} deposited successfully.")
            # Confirmation message
            
        else:
            # If amount is not valid
            print("Invalid deposit amount.")
            # Error message

    def withdraw(self, amount):
        # Method to withdraw money from the account
        if amount > self.balance:
            # Check if withdrawal exceeds available balance
            print("Insufficient balance.")
            # Error message
            
        elif amount <= 0:
            # Check if withdrawal amount is invalid
            print("Invalid withdrawal amount.")
            # Error message
            
        else:
            # If all conditions are satisfied
            self.balance -= amount
            # Deduct amount from balance
            print(f"{amount} withdrawn successfully.")
            # Confirmation message

    def check_balance(self):
        # Method to display account details
        print(f"Account Holder: {self.name}")
        # Print account holder name
        print(f"Current Balance: {self.balance}")
        # Print current balance
```

```python
# Create an object (account)
acc1 = BankAccount("Anish", 1000)

# Call method to check balance
acc1.check_balance()

# Deposit money
acc1.deposit(500)

# Withdraw money
acc1.withdraw(300)

# Check balance again
acc1.check_balance()
```

```python
class LibraryBook:
    
    def __init__(self, title, author):
        # Constructor method to initialize book details
        self.title = title
        # Store the book title
        self.author = author
        # Store the author's name
        self.is_issued = False
        # Track whether the book is issued (default: not issued)

    def issue_book(self):
        # Method to issue the book
        if not self.is_issued:
            # Check if the book is currently available
            self.is_issued = True
            # Mark the book as issued
            print(f"'{self.title}' has been issued.")
            # Confirmation message
            
        else:
            # If the book is already issued
            print(f"'{self.title}' is already issued.")
            # Error message

    def return_book(self):
        # Method to return the book
        if self.is_issued:
            # Check if the book was issued
            self.is_issued = False
            # Mark the book as available
            print(f"'{self.title}' has been returned.")
            # Confirmation message
            
        else:
            # If the book was not issued
            print(f"'{self.title}' was not issued.")
            # Error message

    def get_status(self):
        # Method to check book status
        status = "Issued" if self.is_issued else "Available"
        # Decide status based on boolean value
        print(f"Book: {self.title} | Author: {self.author} | Status: {status}")
        # Display full book information
```

```python
# Create a book object
book1 = LibraryBook("Python Basics", "John Doe")

# Check initial status
book1.get_status()

# Issue the book
book1.issue_book()

# Try issuing again (edge case)
book1.issue_book()

# Return the book
book1.return_book()

# Final status
book1.get_status()
```

```python
class FoodOrder:
    
    def __init__(self, customer_name):
        # Constructor to initialize a new order
        self.customer_name = customer_name
        # Store customer's name
        self.items = []
        # List to store ordered food items
        self.total_amount = 0
        # Track total bill amount
        self.is_delivered = False
        # Track delivery status (default: not delivered)

    def add_item(self, item_name, price):
        # Method to add food items to the order
        self.items.append((item_name, price))
        # Add item as a tuple (name, price)
        self.total_amount += price
        # Update total amount
        print(f"{item_name} added to order.")

    def remove_item(self, item_name):
        # Method to remove an item from the order
        for item in self.items:
            # Loop through all items
            if item[0] == item_name:
                # Check if item name matches
                self.items.remove(item)
                # Remove item from list
                self.total_amount -= item[1]
                # Deduct price from total
                print(f"{item_name} removed from order.")
                return
        
        print(f"{item_name} not found in order.")
        # If item not present

    def view_order(self):
        # Method to display current order
        print(f"Customer: {self.customer_name}")
        # Display customer name
        print("Items:")
        # Header
        for item, price in self.items:
            # Loop through items
            print(f" - {item}: ₹{price}")
            # Display each item
        print(f"Total: ₹{self.total_amount}")
        # Display total bill

    def make_payment(self):
        # Method to simulate payment
        if self.total_amount > 0:
            # Check if there are items in order
            print("Payment successful.")
            # Simulate successful payment
        else:
            print("No items to pay for.")
            # Handle empty order

    def deliver_order(self):
        # Method to mark order as delivered
        if not self.is_delivered:
            # Check if already delivered
            self.is_delivered = True
            # Update delivery status
            print("Order delivered successfully.")
        else:
            print("Order already delivered.")
```

```python
order1 = FoodOrder("Anish")

order1.add_item("Pizza", 250)
order1.add_item("Burger", 150)

order1.view_order()

order1.make_payment()

order1.deliver_order()

order1.deliver_order()
```

# Few More OOP Concepts

```python
# Encapsulation (Data Hiding)

Meaning:
Keeping data safe inside a class and controlling how it is accessed.

Simple Idea:
    “Don’t allow direct access—use methods instead.”
```

```python
class Bank:
    def __init__(self):
        self.__balance = 1000   # private variable

    def get_balance(self):
        return self.__balance
```

```python
Key Points:
    __balance → private (cannot access directly)
    Access through methods → safer
```

```python
# Inheritance (Code Reuse)

Meaning:
One class can reuse properties and methods of another class.

Simple Idea:
    “Child gets everything from parent”
```

```python
class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):
    pass
```

```python
Key Points:
    Dog inherits from Animal
    No need to rewrite speak()
```

```python
# Polymorphism (Many Forms)

Meaning:
Same method name → different behavior

Simple Idea:

“Same action, different result”
```

```python
class Dog:
    def sound(self):
        print("Bark")

class Cat:
    def sound(self):
        print("Meow")
```

```python
Key Points:
    Same method sound()
    Different outputs
```

```python
# Abstraction (Hiding Complexity)

Meaning:
Show only essential features, hide internal logic.

Simple Idea:
    “User clicks button, system does complex work behind”
```

```python
class Payment:
    def pay(self):
        print("Processing payment...")
```

```python
# (User doesn’t see how payment actually happens)
```

```python
| Concept       | Purpose         | Key Idea                   |
| ------------- | --------------- | -------------------------- |
| Encapsulation | Data protection | Hide variables             |
| Inheritance   | Code reuse      | Parent → Child             |
| Polymorphism  | Flexibility     | Same method, diff behavior |
| Abstraction   | Simplicity      | Hide complexity            |


Encapsulation → Security (banking apps)
Inheritance → Code reuse (large systems)
Polymorphism → Flexible design
Abstraction → Clean APIs    
```

```python
class BankAccount:
    
    def __init__(self, balance):
        self.__balance = balance   # private variable

    def deposit(self, amount):
        self.__balance += amount   # controlled update

    def get_balance(self):
        return self.__balance      # controlled access
```

```python
#  Encapsulation (base class)

class Order:
    
    def __init__(self):
        self.__total = 0   # private variable

    def add_amount(self, amount):
        self.__total += amount

    def get_total(self):
        return self.__total

# Only controlled access to total amount
```

```python
# adding inheritence

class Order:
    
    def __init__(self):
        self._total = 0   # protected (accessible to child)

    def add_amount(self, amount):
        self._total += amount


class DeliveryOrder(Order):
    
    def __init__(self):
        super().__init__()
        self.delivery_fee = 50

    def get_total(self):
        return self._total + self.delivery_fee

# DeliveryOrder reuses Order
```

```python
# Adding Polymorphism

class PickupOrder(Order):
    
    def get_total(self):
        return self._total   # no extra charge


# Polymorphism
orders = [DeliveryOrder(), PickupOrder()]

for o in orders:
    o.add_amount(200)
    print(o.get_total())
    
#Same method get_total() → different results
```

```python
# Combine

class Order:
    
    def __init__(self, customer):
        self.customer = customer
        self._total = 0

    def add_item(self, price):
        self._total += price


class DeliveryOrder(Order):
    
    def __init__(self, customer):
        super().__init__(customer)
        self.delivery_fee = 50

    def get_total(self):
        return self._total + self.delivery_fee


class PickupOrder(Order):
    
    def get_total(self):
        return self._total


```

```python
# ---- MAIN PROGRAM ----

o1 = DeliveryOrder("Anish")
o1.add_item(200)

o2 = PickupOrder("John")
o2.add_item(200)

orders = [o1, o2]

for o in orders:
    print(f"{o.customer} pays ₹{o.get_total()}")
```

# Full Banking System

```python
class BankAccount:
    """
    Base class for all types of bank accounts
    Demonstrates Encapsulation + acts as Parent (Inheritance)
    """

    def __init__(self, account_holder, initial_balance):
        self.account_holder = account_holder
        
        self.__balance = initial_balance  
        # PRIVATE variable → Encapsulation (cannot access directly)

    def deposit(self, amount):
        # Add money to account
        
        if amount > 0:
            self.__balance += amount
            print(f"₹{amount} deposited successfully.")
        else:
            print("Invalid deposit amount.")

    def withdraw(self, amount):
        # Basic withdrawal logic
        
        if amount <= 0:
            print("Invalid amount.")
        
        elif amount > self.__balance:
            print("Insufficient balance.")
        
        else:
            self.__balance -= amount
            print(f"₹{amount} withdrawn successfully.")

    def get_balance(self):
        # Controlled access to private balance
        
        return self.__balance

    def calculate_interest(self):
        """
        This method will be overridden by child classes
        Demonstrates Polymorphism
        """
        return 0
```

```python
# Savings Account (Inheritance + Polymorphism)
```

```python
class SavingsAccount(BankAccount):
    """
    Inherits from BankAccount
    Adds interest feature
    """

    def __init__(self, account_holder, balance):
        super().__init__(account_holder, balance)
        self.interest_rate = 0.04   # 4% interest

    def calculate_interest(self):
        # Polymorphism → different logic
        
        return self.get_balance() * self.interest_rate
```

```python
# Current Account (Different Behavior)
```

```python
class CurrentAccount(BankAccount):
    """
    Current account with overdraft facility
    """

    def __init__(self, account_holder, balance):
        super().__init__(account_holder, balance)
        self.overdraft_limit = 1000

    def withdraw(self, amount):
        # Overriding withdrawal → Polymorphism
        
        if amount <= 0:
            print("Invalid amount.")
        
        elif amount > self.get_balance() + self.overdraft_limit:
            print("Overdraft limit exceeded.")
        
        else:
            # Accessing protected behavior via method
            new_balance = self.get_balance() - amount
            
            # Hacky but safe way (since balance is private)
            self._BankAccount__balance = new_balance
            
            print(f"₹{amount} withdrawn (with overdraft if needed).")

    def calculate_interest(self):
        # No interest for current account
        
        return 0
```

```python
# Transaction Class (Encapsulation Example)
class Transaction:
    """
    Handles transaction details
    """

    def __init__(self, amount, type):
        self.__amount = amount   # private
        self.__type = type       # private

    def get_details(self):
        return f"{self.__type} of ₹{self.__amount}"
```

```python
# Main Program

# Create accounts
acc1 = SavingsAccount("Anish", 5000)
acc2 = CurrentAccount("John", 2000)

# Deposit
acc1.deposit(1000)

# Withdraw
acc2.withdraw(2500)   # uses overdraft

# Check balances
print(f"Anish Balance: ₹{acc1.get_balance()}")
print(f"John Balance: ₹{acc2.get_balance()}")

# Interest calculation (Polymorphism)
accounts = [acc1, acc2]

for acc in accounts:
    print(f"{acc.account_holder} Interest: ₹{acc.calculate_interest()}")

# Transaction example
t1 = Transaction(500, "Deposit")
print(t1.get_details())
```

```python
# in the above if you notice 

class SavingsAccount(BankAccount)
class CurrentAccount(BankAccount)

are Inheritance
```

```python
Polymorphism is being used in

calculate_interest()
withdraw()

Same method → different behavior:

Savings → interest
Current → overdraft
```
