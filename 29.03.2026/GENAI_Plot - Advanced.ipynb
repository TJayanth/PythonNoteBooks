```python
# Line Plot (Most Basic Plot)

Used for continuous trends.
Example: simple function.
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10,100)
y = x**2

plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("x")
plt.ylabel("x^2")
plt.show()
```

```python
# Line plots → trends over time or continuous functions

Examples:

    training loss
    stock price
    sensor data
```

```python
# Scatter Plot

Used for relationship between variables.
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.rand(50)
y = 2*x + np.random.randn(50)*0.2

plt.scatter(x,y)
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
Scatter plots → find relationships

Used in:

    regression
    clustering
    correlation analysis
```

```python
# Bar Chart

Used for categorical data.
```

```python
import matplotlib.pyplot as plt

students = ["A","B","C","D"]
marks = [70,85,60,90]

plt.bar(students,marks)

plt.title("Student Marks")
plt.xlabel("Student")
plt.ylabel("Marks")

plt.show()
```

```python
# Bar charts → compare categories
```

```python
# Histogram

Used for distribution of data.
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(0,1,1000)

plt.hist(data,bins=30)

plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()
```

```python
# Histogram → visualize distributions
```

```python
# Box Plot

Shows median, quartiles, and outliers.
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(0,1,100)

plt.boxplot(data)

plt.title("Box Plot")

plt.show()
```

```python
# Multiple Plots
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10,100)

fig, ax = plt.subplots(2,2, figsize=(8,6))

ax[0,0].plot(x,x)
ax[0,0].set_title("y=x")

ax[0,1].plot(x,x**2)
ax[0,1].set_title("y=x^2")

ax[1,0].plot(x,np.sin(x))
ax[1,0].set_title("sin(x)")

ax[1,1].plot(x,np.cos(x))
ax[1,1].set_title("cos(x)")

plt.tight_layout()
plt.show()
```

```python
# 3D Surface Plot
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5,5,100)
y = np.linspace(-5,5,100)

X,Y = np.meshgrid(x,y)

Z = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X,Y,Z, cmap='viridis')

ax.set_title("3D Surface Plot")

plt.show()
```

```python
Useful in:

optimization
loss surfaces
physics simulations
```

```python
# Correlation plot 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = pd.DataFrame(np.random.rand(10,5),
                    columns=["A","B","C","D","E"])

corr = data.corr()

sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap")
plt.show()
```

```python
# Animated Plot
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

x = np.linspace(0,10,100)

fig, ax = plt.subplots()

line, = ax.plot(x, np.sin(x))

def update(frame):
    line.set_ydata(np.sin(x + frame/10))
    return line,

ani = FuncAnimation(fig, update, frames=100, interval=50)

ani.save("sine_wave.gif", writer="pillow")

plt.show()
```

```python
# Contour Plot

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3,3,100)
y = np.linspace(-3,3,100)

X,Y = np.meshgrid(x,y)

Z = X**2 + Y**2

plt.contour(X,Y,Z)

plt.title("Contour Plot")

plt.show()
```

```python
# A table for someone who wants to know when to use what plot......
```

| Plot Type              | Data Type          | When to Use                                  | What It Shows                   | Example Use Case           |
| ---------------------- | ------------------ | -------------------------------------------- | ------------------------------- | -------------------------- |
| Line Plot              | Continuous         | Show trends over time                        | Change of values across an axis | Training loss in ML        |
| Scatter Plot           | Continuous         | Relationship between variables               | Correlation patterns            | Height vs weight           |
| Bar Chart              | Categorical        | Compare categories                           | Relative magnitude              | Sales by product           |
| Stacked Bar Chart      | Categorical        | Show composition within categories           | Part-to-whole relationships     | Revenue breakdown          |
| Histogram              | Continuous         | Understand distribution                      | Frequency of values             | Age distribution           |
| Box Plot               | Continuous         | Detect outliers & quartiles                  | Median, IQR, outliers           | Salary distribution        |
| Violin Plot            | Continuous         | Compare distributions                        | Density + box plot info         | Compare exam scores        |
| KDE Plot               | Continuous         | Smooth distribution estimate                 | Probability density             | Data distribution shape    |
| Heatmap                | Matrix / Tabular   | Visualize correlations or intensity          | Color-coded values              | Feature correlation matrix |
| Pair Plot              | Multiple variables | Explore relationships between many variables | Scatter matrix                  | Exploratory data analysis  |
| Area Plot              | Continuous         | Show cumulative trends                       | Magnitude over time             | Population growth          |
| Pie Chart              | Categorical        | Show proportions                             | Percent contribution            | Market share               |
| Donut Chart            | Categorical        | Similar to pie, cleaner look                 | Proportion                      | Budget allocation          |
| Contour Plot           | Continuous (2D)    | Visualize surfaces                           | Level curves                    | Optimization landscape     |
| 3D Surface Plot        | Continuous (3D)    | Visualize functions in 3D                    | Surface relationships           | Loss surface in ML         |
| Network Graph          | Graph data         | Show connections                             | Nodes & relationships           | Social networks            |
| Time Series Plot       | Time data          | Analyze temporal patterns                    | Trend, seasonality              | Stock price                |
| Bubble Chart           | Continuous + size  | Add third variable                           | Magnitude with position         | GDP vs life expectancy     |
| Radar Chart            | Multivariate       | Compare multiple metrics                     | Profile comparison              | Model performance metrics  |
| Treemap                | Hierarchical       | Visualize nested proportions                 | Hierarchical structure          | Disk usage                 |
| Parallel Coordinates   | Multivariate       | Compare many variables                       | Patterns across features        | ML feature comparison      |
| Streamgraph            | Time series        | Show evolving categories                     | Flow over time                  | Website traffic sources    |
| Ridge Plot             | Continuous groups  | Compare multiple distributions               | Distribution comparison         | Multiple population groups |
| Hexbin Plot            | Dense scatter data | Avoid overplotting                           | Density of points               | Large datasets             |
| Joint Plot             | Two variables      | Combine scatter + distribution               | Correlation & distribution      | ML feature relationships   |
| Decision Boundary Plot | ML data            | Visualize classifier behavior                | Model separation                | Logistic regression        |


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

np.random.seed(0)

data = np.random.randn(100)
x = np.linspace(0,10,100)
y = np.sin(x)
x2 = np.random.randn(200)
y2 = np.random.randn(200)

df = pd.DataFrame({
    "A": np.random.randn(100),
    "B": np.random.randn(100),
    "C": np.random.randn(100)
})

fig, axs = plt.subplots(4,4, figsize=(16,16))

# 1 Line Plot
axs[0,0].plot(x,y)
axs[0,0].set_title("Line Plot")

# 2 Scatter Plot
axs[0,1].scatter(x2,y2)
axs[0,1].set_title("Scatter Plot")

# 3 Bar Chart
axs[0,2].bar(["A","B","C"],[5,7,3])
axs[0,2].set_title("Bar Chart")

# 4 Histogram
axs[0,3].hist(data,bins=20)
axs[0,3].set_title("Histogram")

# 5 Box Plot
axs[1,0].boxplot(data)
axs[1,0].set_title("Box Plot")

# 6 Violin Plot
sns.violinplot(data=data, ax=axs[1,1])
axs[1,1].set_title("Violin Plot")

# 7 KDE Plot
sns.kdeplot(data, ax=axs[1,2])
axs[1,2].set_title("KDE Plot")

# 8 Heatmap
sns.heatmap(df.corr(), ax=axs[1,3])
axs[1,3].set_title("Heatmap")

# 9 Hexbin
axs[2,0].hexbin(x2,y2, gridsize=20)
axs[2,0].set_title("Hexbin Plot")

# 10 Area Plot
axs[2,1].fill_between(x,y)
axs[2,1].set_title("Area Plot")

# 11 Pie Chart
axs[2,2].pie([30,40,30], labels=["A","B","C"])
axs[2,2].set_title("Pie Chart")

# 12 Bubble Chart
sizes = np.random.rand(50)*300
axs[2,3].scatter(np.random.rand(50), np.random.rand(50), s=sizes)
axs[2,3].set_title("Bubble Chart")

# 13 Contour Plot
X,Y = np.meshgrid(np.linspace(-3,3,100), np.linspace(-3,3,100))
Z = X**2 + Y**2
axs[3,0].contour(X,Y,Z)
axs[3,0].set_title("Contour Plot")

# 14 Density Heatmap
sns.histplot(x=x2, y=y2, ax=axs[3,1])
axs[3,1].set_title("2D Density")

# 15 Step Plot
axs[3,2].step(x,y)
axs[3,2].set_title("Step Plot")

# 16 Stem Plot
axs[3,3].stem(np.arange(10), np.random.rand(10))
axs[3,3].set_title("Stem Plot")

plt.tight_layout()
plt.show()
```
