# Some Object-Oriented Programming (OOP) concepts which u already know

```python

# Basics: Classes, Objects, Attributes, Methods

class Car:# This defines a class called Car.
    """A simple Car class.""" 
#     We add self in Python classes because it refers to the current instance of the class. 
#     It’s how an object keeps track of its own data and methods.
    def __init__(self, make, model, year): #→ This is the constructor method that initializes a new Car object.
        self.make = make # Public Attribute
        self.model = model # Public Attribute
        self.year = year # Public Attribute
 
    def display_info(self):# This is a method, a function that belongs to the class and acts on its data.
        """Displays car information."""
        print(f"{self.year} {self.make} {self.model}")

# Creating an object
car1 = Car("Toyota", "Corolla", 2020) #car1 is an object, an instance of the Car class.
car1.display_info()#This calls the display_info method on the car1 object.
```

```python
# For understanding
class Student:
    """A simple Student class."""
    # The self parameter refers to the current instance of the class.
    def __init__(self, name, age, grade):
        self.name = name  # Public Attribute
        self.age = age    # Public Attribute
        self.grade = grade # Public Attribute

    def display_info(self):
        """Displays student information."""
        print(f"Name: {self.name}, Age: {self.age}, Grade: {self.grade}")

# Creating an object
student1 = Student("Alice", 14, "8th Grade")
student1.display_info()  # Calls the display_info method on the student1 object
```

```python
# For understanding
class Book:
    """A class representing a book in a library."""

    def __init__(self, title, author, year):
        self.title = title  # Public Attribute
        self.author = author  # Public Attribute
        self.year = year  # Public Attribute
        self._is_borrowed = False  # Protected Attribute

    def display_info(self):
        """Displays book information."""
        status = 'Borrowed' if self._is_borrowed else 'Available'
        print(f"{self.title} by {self.author}, {self.year} - Status: {status}")

    def borrow(self):
        """Marks the book as borrowed."""
        if not self._is_borrowed:
            self._is_borrowed = True
            print(f"You have borrowed '{self.title}'.")
        else:
            print(f"'{self.title}' is already borrowed.")

    def return_book(self):
        """Marks the book as returned."""
        if self._is_borrowed:
            self._is_borrowed = False
            print(f"You have returned '{self.title}'.")
        else:
            print(f"'{self.title}' was not borrowed.")


# book objects
book1 = Book("1984", "George Orwell", 1949)
book2 = Book("To Kill a Mockingbird", "Harper Lee", 1960)

# Displaying book info
book1.display_info()
book2.display_info()

# Borrowing and returning books
book1.borrow()
book1.display_info()
book1.return_book()
book1.display_info()

book2.borrow()
book2.borrow()  # Trying to borrow again
book2.return_book()
```

```python
# access the main class - CAR

# Inheritance - concept
class ElectricCar(Car):#This defines a subclass of Car, using inheritance.
    """Represents an electric car, a subclass of Car."""
    def __init__(self, make, model, year, battery_size):##→ This is the constructor method that initializes for this ElectricCar.
        super().__init__(make, model, year)#Calls the parent class’s constructor from the above cell.
        self.battery_size = battery_size #An additional attribute specific to ElectricCar (this subclass).

    def display_info(self): # method
        """Overrides the parent method to include battery size."""
        print(f"{self.year} {self.make} {self.model} with {self.battery_size}-kWh battery")

# Polymorphism
cars = [Car("Honda", "Civic", 2019), ElectricCar("Tesla", "Model 3", 2021, 75)]
#Both are objects, demonstrating polymorphism when calling display_info().

for car in cars:
    car.display_info()  # Calls the right version of display_info for each car

```

```python
# Encapsulation - concept
class BankAccount:#Defines the BankAccount class.
    """A simple bank account class with encapsulation."""
    def __init__(self, owner, balance):
        self.owner = owner # public attribute
        self.__balance = balance  # A private attribute (mangling), meant to be hidden from direct external access.

    def deposit(self, amount):#These are methods that control access to the private balance attribute.
        if amount > 0:
            self.__balance += amount
            print(f"Deposited ${amount}. New balance: ${self.__balance}")

    def withdraw(self, amount):#These are methods that control access to the private balance attribute.
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew ${amount}. New balance: ${self.__balance}")
        else:
            print("Insufficient funds or invalid amount.")

# Using the BankAccount class
account = BankAccount("Alice", 1000)#account is an object.
account.deposit(500)#Calls the deposit method on the account object.
account.withdraw(300)#Calls the withdraw method on the account object.

# Trying to access a private attribute (will raise an error)
#print(account.__balance)
print(account._BankAccount__balance)  # Accessing the private attribute directly (not recommended)

# A double underscore (__) triggers name mangling — Python changes the name of the attribute
# to _ClassName__attribute, making it harder to access directly.
#That’s why self.__balance becomes _BankAccount__balance.


```

```python
# Employee Class
class Employee:
    """A simple Employee class."""
    def __init__(self, name, employee_id, salary):
        self.name = name               # Public attribute
        self.employee_id = employee_id # Public attribute
        self.__salary = salary         # Private attribute

    def display_info(self):
        """Displays employee information."""
        print(f"Employee: {self.name}, ID: {self.employee_id}")

    def give_raise(self, amount):
        """Gives a raise to the employee."""
        if amount > 0:
            self.__salary += amount
            print(f"{self.name} got a raise! New salary: ${self.__salary}")

    def get_salary(self): # a seperate method to acess the salary
        """Public method to safely access the private salary."""
        return self.__salary

# Using the Employee class
employee1 = Employee("Bob", 12345, 50000)
employee1.display_info()
employee1.give_raise(5000)
print(f"Salary: ${employee1.get_salary()}")

# Accessing private attribute directly (not recommended)
print(f"(Accessing private attribute directly) Salary: ${employee1._Employee__salary}")
```

```python
# An example of Inheritance and Encapsulation

class Animal:
    """Base class representing a general animal."""
    def __init__(self, name, age):
        self.name = name  # Public attribute
        self._age = age  # Protected attribute

    def display_info(self):
        """Displays animal information."""
        print(f"Name: {self.name}, Age: {self._age}")

class Dog(Animal):
    """Derived class representing a dog."""
    def __init__(self, name, age, breed):
        super().__init__(name, age)
        self.breed = breed  # Public attribute

    def display_info(self):
        """Overrides the base class method to add breed information."""
        super().display_info()
        print(f"Breed: {self.breed}")

# Creating an object of the Dog class
dog = Dog("Buddy", 5, "Golden Retriever")
dog.display_info()
```

# Error Handling

```python
# Error Handling and Exceptions in Python

# Basic Try-Except Block
try:
    number = int(input("Enter a number: ")) # input a built-in Python function 
    print(f"You entered: {number}")
except ValueError:
    print("Invalid input! Please enter a valid number.")


```

```python
# Handling Multiple Exceptions
try:
    num1 = int(input("Enter first number: "))
    num2 = int(input("Enter second number: "))
    result = num1 / num2
    print(f"Result: {result}")
except ValueError:
    print("Please enter valid numbers!")
except ZeroDivisionError:
    print("Division by zero is not allowed!")
```

```python
#  Raising Exceptions
def set_age(age):
    if age < 0:
        raise ValueError("Age cannot be negative!")
    print(f"Age set to {age}")

    
try:
    set_age(-5)
except ValueError as e:
    print(e)

# raise is telling Python to stop execution and throw an error.
# ValueError is the type of exception being raised.
```

```python
# Custom Exceptions
class NegativeBalanceError(Exception):
    """Custom exception for negative balance."""
    pass

def withdraw(balance, amount):
    if amount > balance:
        raise NegativeBalanceError("Insufficient balance!")
    balance -= amount
    print(f"Withdrawn: ${amount}. Remaining balance: ${balance}")

try:
    withdraw(100, 200)
except NegativeBalanceError as e:
    print(e)
```

```python
# an use case example
class FileProcessor:
    """A class to handle file reading and processing with robust error handling."""
    def __init__(self, filename):
        self.filename = filename

    def read_file(self):
        """Reads the content of a file and handles various exceptions."""
        try:
            with open(self.filename, 'r') as file:
                content = file.read()
                print("File content successfully read.")
                return content
        except FileNotFoundError:
            print(f"Error: The file '{self.filename}' was not found.")
        except PermissionError:
            print(f"Error: Permission denied for file '{self.filename}'.")
        except Exception as e:
            print(f"An unexpected error occurred: {e}")
        finally:
            print("File reading attempt finished.")

    def process_content(self):
        """Processes the content of the file if it exists."""
        content = self.read_file()
        if content:
            lines = content.splitlines()
            print(f"The file has {len(lines)} lines.")

# Using the FileProcessor class
processor = FileProcessor("requirements1.txt")
processor.process_content()
```
