# Basics of Probability - a crude way to remember things out

```python
# Probability of rolling a 6 on a fair die

favorable_outcomes = 1
total_outcomes = 6
prob_rolling_6 = favorable_outcomes / total_outcomes

print(f"Probability of rolling a 6: {prob_rolling_6:.4f}")
```

```python
# 🔹 Monte Carlo Simulation: Rolling a Die 10,000 Times

import numpy as np

# Simulate rolling a die 10,000 times
np.random.seed(42)  # For reproducibility
rolls = np.random.randint(1, 7, size=10000)  # Generate numbers from 1 to 6

# Probability of rolling a 6
prob_rolling_6_simulated = np.sum(rolls == 6) / len(rolls)

print(f"Simulated Probability of rolling a 6: {prob_rolling_6_simulated:.4f}")
```

# Probability Mass Function (PMF)

```python
# Probability Mass Function (PMF)

# The Probability Mass Function (PMF) is used for discrete random variables (e.g., dice rolls, coin flips).
# It assigns a probability to each possible outcome.

#Example: PMF of a Fair Die

import matplotlib.pyplot as plt
import seaborn as sns

# Count occurrences of each number (1-6)
values, counts = np.unique(rolls, return_counts=True)

# Compute PMF
pmf = counts / len(rolls)

# Plot PMF
plt.bar(values, pmf, color='skyblue', edgecolor='black')
plt.xlabel("Die Face")
plt.ylabel("Probability")
plt.title("Probability Mass Function (PMF) of a Fair Die")
plt.show()

# Each face 1,2,3,4,5,6 has ~1/6 probability.
# PMFs only apply to discrete values.
    
```

# Probability Density Function (PDF)

```python
# A Probability Density Function (PDF) describes continuous variables (e.g., heights, weights).
# Unlike PMFs, PDFs don’t assign probabilities to exact values but to ranges.
```

```python
 # Example: PDF of Heights
from scipy.stats import norm

# Generate normal distribution (mean=170 cm, std=10 cm)
x = np.linspace(140, 200, 1000)
pdf = norm.pdf(x, loc=170, scale=10)

# Plot PDF
plt.plot(x, pdf, color='blue', label="PDF")
plt.fill_between(x, pdf, alpha=0.2, color='blue')
plt.xlabel("Height (cm)")
plt.ylabel("Density")
plt.title("Probability Density Function (PDF) of Heights")
plt.legend()
plt.show()

# The peak shows the most likely height (~170 cm).
# Shorter & taller heights occur less frequently.
                                           
```

# Cumulative Distribution Function (CDF)


```python
# The Cumulative Distribution Function (CDF) gives the probability that a variable is less than or equal to a given value.
```

```python
#Example: What’s the probability that a person is shorter than 175 cm?

# Compute CDF
cdf = norm.cdf(x, loc=170, scale=10)

# Plot CDF
plt.plot(x, cdf, color='red', label="CDF")
plt.xlabel("Height (cm)")
plt.ylabel("Probability")
plt.title("Cumulative Distribution Function (CDF) of Heights")
plt.legend()
plt.show()

# Probability of height ≤ 175 cm
p_175 = norm.cdf(175, loc=170, scale=10)
print(f"Probability of height ≤ 175 cm: {p_175:.2%}")

# CDF helps answer probability of ranges (e.g., "below 175 cm").
```

# Conditional Probability

```python
# Conditional probability measures the probability of an event given another event has occurred.

# 🔹 Example:

#     Event A: The student passes an exam.
#     Event B: The student studied for the exam.
#     We want to find P(A | B) → Probability of passing given they studied.

# P(A∣B)=P(A∩B)P(B)

# Given probabilities
P_Study = 0.6  # 60% students study
P_Pass_given_Study = 0.9  # 90% pass if they study
P_Pass_given_NoStudy = 0.2  # 20% pass if they don’t study

# Total probability of passing
P_Pass = (P_Pass_given_Study * P_Study) + (P_Pass_given_NoStudy * (1 - P_Study))

# Applying Bayes' Theorem
P_Study_given_Pass = (P_Pass_given_Study * P_Study) / P_Pass

print(f"Probability of studying given the student passed: {P_Study_given_Pass:.2%}")
```

# Expected Value and Variance

```python
# Expected Value (Mean): Long-term average outcome of a random variable.
# Variance: Measures spread (how far values deviate from the mean).

# # E(X)=∑xP(x)
# # Var(X)=E(X2)−(E(X))2

# Example: Expected Value & Variance of a Die

# Possible outcomes
x = np.array([1, 2, 3, 4, 5, 6])
p = np.ones(6) / 6  # Equal probability

# Expected Value
expected_value = np.sum(x * p)

# Variance
variance = np.sum((x**2) * p) - expected_value**2

print(f"Expected Value (Mean) = {expected_value:.2f}")
print(f"Variance = {variance:.2f}")

# The average roll of a die is 3.5, even though 3.5 never actually occurs!
```

# Normal Distribution

```python

import numpy as np
import matplotlib.pyplot as plt
import math

# Function to calculate normal distribution PDF
def normal_pdf(x, mu, sigma):
    coefficient = 1 / (sigma * math.sqrt(2 * math.pi))
    exponent = math.exp(-((x - mu) ** 2) / (2 * sigma ** 2))
    return coefficient * exponent

# Parameters
mu = 0      # Mean
sigma = 1   # Standard Deviation

# Generate x values
x_values = np.linspace(-4, 4, 100)
y_values = [normal_pdf(x, mu, sigma) for x in x_values]

# Plot the normal distribution
plt.plot(x_values, y_values, label='Normal Distribution (μ=0, σ=1)', color='red')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.title('Normal Distribution')
plt.legend()
plt.grid()
plt.show()
```
