```python
## Functions and Modular Programming in Python

# Basics: Writing Functions

def greet(name):
    """This function greets the user by name."""
    print(f"Hello, {name}!")

greet("Everyone")
```

```python
# Parameters, Return Values, and Scope

def add_numbers(a, b):
    result = a + b
    return result

sum_result = add_numbers(5, 3)
print(f"Sum: {sum_result}")
```

```python
# Default Parameters

def power(base, exponent=2):
    return base ** exponent

print(power(3))        # Uses default exponent
print(power(3,3))     # Overrides default
```

```python
# *args and **kwargs
#In Python, *args and **kwargs are used to allow functions to accept an arbitrary number of arguments.

def flexible_function(*args1, **kwargs):
    print("Positional arguments:", args1)
    print("Keyword arguments:", kwargs)

flexible_function(1, 2, 3, name="Alice", age=25)

```

```python
# Lambda Functions
#A lambda function in Python is a small, anonymous function — meaning it doesn’t have a name like 
# regular functions defined with def. 
# It’s often used for short, simple operations.

square = lambda x: x ** 2
print(f"Square of 4: {square(4)}")

# creates a function that takes one argument (x) and returns its square (x ** 2). 
# It’s the same as writing:

# def square(x):
#     return x ** 2


# So why do we use this ?

# The main reason for using lambdas is when you need a quick function without the overhead of defining it with def.
```

```python
# Decorators (Advanced)

# This is the decorator function. 
# It takes another function (func) as an argument — that’s key because decorators “wrap” other functions.

def decorator_example(func):
    def wrapper():
        print("Before the function call")
        func()
        print("After the function call")
    return wrapper

@decorator_example
def say_hello():
    print("Hello!")
    
@decorator_example
def say_WhatUrdoing():
    print("WhatUrdoing!")
    
# The @decorator_example applies the decorator to say_hello(). 
# So when you call say_hello(), Python actually calls wrapper() from the decorator.

say_hello()
say_WhatUrdoing()
```

```python
import time

#use case. You can measure latency/time taken of any function through this

def timing_decorator(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)# the actual function where it being called
        end_time = time.time()
        print(f"Execution time: {end_time - start_time:.4f} seconds")
        return result
    return wrapper

@timing_decorator
def compute_sum(n):
    return sum(range(n))

# Compute the sum of the first million numbers
print(compute_sum(1000000))
```

```python
# Recursive Functions

def factorial(n):
    if n == 0:
        return 1
    else:
        #print(n * factorial(n-1))
        return  n * factorial(n-1)

print(f"Factorial of 5: {factorial(5)}")


# if you dont understand the above you can use this, try starting from (1)
# def factorial(n):
#     print(f"factorial called with n={n}")
#     if n == 0:
#         print("Base case reached, returning 1")
#         return 1
#     else:
#         result = n * factorial(n-1)
#         print(f"Returning {result} for factorial({n})")
#         return result
    
    
# print(f"Factorial of 5: {factorial(5)}")   
```

```python
# Higher-Order Functions: map, filter, reduce
from functools import reduce

numbers = [1, 2, 3, 4, 5]

# map: Apply a function to each element
squares = list(map(lambda x: x ** 2, numbers))
print(f"Squares: {squares}")

# filter: Keep elements that match a condition
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(f"Even numbers: {even_numbers}")

# reduce: Combine elements into a single result
product = reduce(lambda x, y: x * y, numbers)
print(f"Product of numbers: {product}")
```

```python
# Error Handling in Functions

#print(5/0)# Run this First

def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Error: Division by zero!"
        
print(safe_divide(10, 3))
print(safe_divide(5, 0))
```

```python
# Using Modules and Libraries
import math
import random

print(f"Square root of 16: {math.sqrt(16)}")
print(f"Random number between 1 and 100: {random.randint(1, 100)}")
```

```python
# Writing Docstrings and Comments -> Good Practice

def calculate_area(length, width):
    """Calculates the area of a rectangle.

    Args:
        length (float): The length of the rectangle.
        width (float): The width of the rectangle.

    Returns:
        float: The calculated area of the rectangle.

    Example:
        >>> calculate_area(5, 3)
        15
    """
    return length * width

# Example usage
print(f"Area of rectangle: {calculate_area(5, 3)}")
```
