# Parallel Processing in Python

```python
#Python has built-in libraries like multiprocessing for parallel execution
```

```python
import multiprocessing

def worker(num):
    print(f"Worker {num}")

if __name__ == "__main__":
    processes = []
    for i in range(4):  # Create 4 parallel processes
        p = multiprocessing.Process(target=worker, args=(i,))
        processes.append(p)
        p.start()

    for p in processes:
        p.join()
```

```python
# Parallelizing CPU-Intensive Tasks
```

```python
import multiprocessing

def square(n):
    return n * n

if __name__ == "__main__":
    nums = [1, 2, 3, 4, 5]
    with multiprocessing.Pool(processes=3) as pool:
        results = pool.map(square, nums)
    print(results) 
```

```python
# Parallel Processing with concurrent.futures
```

```python
import concurrent.futures

def task(n):
    return n * n

if __name__ == "__main__":
    with concurrent.futures.ProcessPoolExecutor(max_workers=3) as executor:
        results = executor.map(task, [1, 2, 3, 4, 5])
    print(list(results))
```

```python
# Let’s say you have a large number of images and you want to apply a transformation (e.g., resizing, converting to grayscale, 
# or applying filters).
# Doing this sequentially would be slow. Using parallel processing can significantly speed up the task.

import os
import cv2
import multiprocessing

def convert_to_grayscale(image_path):
    """Reads an image, converts it to grayscale, and saves it."""
    img = cv2.imread(image_path)
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    output_path = f"grayscale_{os.path.basename(image_path)}"
    cv2.imwrite(output_path, gray_img)
    print(f"Processed: {output_path}")

if __name__ == "__main__":
    image_folder = "images/"
    image_files = [os.path.join(image_folder, f) for f in os.listdir(image_folder) if f.endswith(".jpg")]

    with multiprocessing.Pool(processes=4) as pool:  # Use 4 CPU cores
        pool.map(convert_to_grayscale, image_files)

    print("All images processed successfully!")


# If you process many images sequentially, it could take several minutes.
# Using multiprocessing, it divides the workload among multiple CPU cores, reducing processing time significantly.
```

```python
# Web Scraping in Parallel
# Web scraping is the automated process of extracting data from websites, using software (often called "bots" or "scrapers") 
# to fetch and parse the underlying HTML code and data, which can then be stored or used for various purposes

# If you need to scrape multiple web pages, doing it sequentially would be slow. 
# Using parallel processing allows multiple pages to be scraped at the same time.

import requests
import concurrent.futures

urls = [
    "https://www.hindustantimes.com/india-news/lashkaretaiba-terrorist-abu-qatal-mastermind-of-rajouri-attack-shot-dead-in-pakistan-report-101742090607318.html",
    "https://www.thehindu.com/news/cities/Delhi/delhi-records-cleanest-air-quality-in-three-years-with-aqi-of-85/article69336033.ece",
]

def fetch_url(url):
    response = requests.get(url)
    print(f"Fetched {url} with status: {response.status_code}")

if __name__ == "__main__":
    with concurrent.futures.ThreadPoolExecutor(max_workers=4) as executor:
        executor.map(fetch_url, urls)
```

```python
import os
import cv2
import multiprocessing
import time
import numpy as np

# Function to convert an image to grayscale
def convert_to_grayscale(image_path):
    img = cv2.imread(image_path)
    if img is None:
        return  # Skip if image is not loaded
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    output_path = f"grayscale_{os.path.basename(image_path)}"
    cv2.imwrite(output_path, gray_img)

# Generate dummy images for testing
def generate_dummy_images(folder, num_images=100):
    if not os.path.exists(folder):
        os.makedirs(folder)
    for i in range(num_images):
        img = (255 * np.random.rand(1000, 1000, 3)).astype(np.uint8)  # Large images
        cv2.imwrite(os.path.join(folder, f"image_{i}.jpg"), img)

# Benchmark Sequential Processing
def process_sequentially(image_files):
    start_time = time.time()
    for img in image_files:
        convert_to_grayscale(img)
    end_time = time.time()
    return end_time - start_time

# Benchmark Parallel Processing
def process_parallel(image_files, num_workers=4):
    start_time = time.time()
    with multiprocessing.Pool(processes=num_workers) as pool:
        pool.map(convert_to_grayscale, image_files)
    end_time = time.time()
    return end_time - start_time

# Main execution
if __name__ == "__main__":
    image_folder = "images/"
    
    # Generate test images (if not already created)
    generate_dummy_images(image_folder, num_images=100)

    # Get image file paths
    image_files = [os.path.join(image_folder, f) for f in os.listdir(image_folder) if f.endswith(".jpg")]

    # Run sequential processing
    time_seq = process_sequentially(image_files)

    # Run parallel processing
    time_par = process_parallel(image_files, num_workers=4)

    # Print results
    print(f"Sequential Processing Time: {time_seq:.4f} seconds")
    print(f"Parallel Processing Time: {time_par:.4f} seconds")
    print(f"Speedup Factor: {time_seq / time_par:.2f}x")
```

# Parallel Matrix Computation (Using NumPy & Joblib)

```python
# Sequential Approach
import numpy as np

A = np.random.rand(5000, 5000)
B = np.random.rand(5000, 5000)

result = np.dot(A, B)  # Single-threaded
print(result)
```

```python
# parallel processing
from joblib import Parallel, delayed

def parallel_dot_product(i):
    return np.dot(A[i], B)

result_parallel = Parallel(n_jobs=4)(delayed(parallel_dot_product)(i) for i in range(A.shape[0]))
print(np.array(result_parallel))
```

# Parallel Sorting (Using multiprocessing)

```python
import numpy as np
import time

arr = np.random.randint(0, 1000000, size=100000000)
start = time.time()
sorted_arr = sorted(arr)  #Python's built-in sorted() function is highly optimized (Timsort) and often runs faster than parallel sorting.
end = time.time()
print(f"Time taken: {end - start:.2f} sec")
```

```python
from multiprocessing import Pool

def parallel_sort(chunk):
    return sorted(chunk)

arr_chunks = np.array_split(arr, 4)  # Split array into 4 chunks
with Pool(4) as p:
    sorted_chunks = p.map(parallel_sort, arr_chunks)

sorted_arr_parallel = np.concatenate(sorted_chunks)
print(f"Parallel Sorting Time: {time.time() - start:.2f} sec")
```

# Tensor with parallel processing 

```python
# Example 1

import torch
import time
from joblib import Parallel, delayed

# Set device (use GPU if available)
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")

# Create large random tensors
size = 100  # 10,000 x 10,000 matrix
A = torch.rand(size, size)
B = torch.rand(size, size)

### **Sequential Matrix Multiplication (CPU)**
start_time = time.time()
result_seq = torch.mm(A, B)  # Single-threaded
end_time = time.time()
print(f"Sequential CPU Time: {end_time - start_time:.4f} sec")

### **2️Parallel Matrix Multiplication (Multi-Core CPU)**
torch.set_num_threads(8)  # Set to 8 CPU threads (adjust as per CPU)
start_time = time.time()
result_parallel = torch.mm(A, B)  # Multi-threaded CPU
end_time = time.time()
print(f"Parallel CPU Time (8 cores): {end_time - start_time:.4f} sec")

### **GPU Accelerated Matrix Multiplication (if GPU available)**
if torch.cuda.is_available():
    A_gpu = A.to(device)
    B_gpu = B.to(device)
    
    start_time = time.time()
    result_gpu = torch.mm(A_gpu, B_gpu)  # GPU computation
    torch.cuda.synchronize()  # Ensure all GPU operations are done
    end_time = time.time()
    
    print(f"GPU Time: {end_time - start_time:.4f} sec")
```

```python
# Example 2

import torch
import time
from multiprocessing import Pool

# Check for GPU
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")

# Create large tensors
size = 100  # 100 million elements
A = torch.rand(size)
B = torch.rand(size)

### **Sequential Addition (CPU)**
start_time = time.time()
C_seq = A + B  # Element-wise addition
end_time = time.time()
print(f"Sequential CPU Time: {end_time - start_time:.4f} sec")

### **Multi-core Parallel Processing (CPU)**
def parallel_add(start, end):
    return A[start:end] + B[start:end]

num_cores = 8  # Adjust based on CPU cores
chunk_size = size // num_cores
indices = [(i * chunk_size, (i + 1) * chunk_size) for i in range(num_cores)]

start_time = time.time()
with Pool(num_cores) as pool:
    results = pool.starmap(parallel_add, indices)

C_parallel = torch.cat(results)
end_time = time.time()
print(f"Parallel CPU Time (8 cores): {end_time - start_time:.4f} sec")

### **GPU Processing**
if torch.cuda.is_available():
    A_gpu = A.to(device)
    B_gpu = B.to(device)

    start_time = time.time()
    C_gpu = A_gpu + B_gpu
    torch.cuda.synchronize()  # Ensure all GPU operations are completed
    end_time = time.time()

    print(f"GPU Time: {end_time - start_time:.4f} sec")
```

```python
# some experiements to show parallel processing

import multiprocessing
import time
import random

# Simulate a CPU-heavy task (racecar)
def racecar(name):
    total_distance = 100
    distance = 0

    while distance < total_distance:
        time.sleep(random.uniform(0.1, 0.3))  # Random delay
        distance += random.randint(5, 15)  # Random speed
        print(f"{name} is at {distance}/{total_distance}")

    print(f"{name} finished the race!")

if __name__ == "__main__":
    cars = ["Ferrari", "Lamborghini", "Tesla", "Bugatti"]
    processes = [multiprocessing.Process(target=racecar, args=(car,)) for car in cars]

    for p in processes:
        p.start()

```

```python
import cv2
import multiprocessing
import numpy as np
import time

# Function to access webcam and display video
def webcam_feed():
    cap = cv2.VideoCapture(0)  # Open webcam (0 for default camera)
    
    if not cap.isOpened():
        print("Error: Could not access webcam")
        return
    
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        
        cv2.imshow("Webcam Feed", frame)
        
        # Press 'q' to exit
        if cv2.waitKey(30) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()

# Function for another computation
def heavy_computation():
    print("Starting heavy computation...")
    start_time = time.time()
    
    # Generating large random numbers and summing them
    arr = np.random.rand(1000)  # 100 million random numbers
    total_sum = np.sum(arr)
    
    end_time = time.time()
    print(f"Heavy Computation Done! Sum: {total_sum:.2f}")
    print(f"Time Taken: {end_time - start_time:.2f} seconds")

# Main function to run both tasks in parallel
if __name__ == "__main__":
    # Creating two processes
    webcam_process = multiprocessing.Process(target=webcam_feed)
    computation_process = multiprocessing.Process(target=heavy_computation)
    
    # Start both processes
    webcam_process.start()
    computation_process.start()
    
    # Wait for both to complete
    webcam_process.join()
    computation_process.join()

    print("Both processes completed.")
```

```python
# Parallel Webcam Capture & Grayscale Conversion

import cv2
import os
import time
import multiprocessing

# Create a folder to store images
FOLDER_NAME = "GENAI"
os.makedirs(FOLDER_NAME, exist_ok=True)

# Function to capture images from webcam and save them
def capture_images():
    cap = cv2.VideoCapture(0)  # Open webcam (0 for default camera)
    
    if not cap.isOpened():
        print("Error: Could not access webcam")
        return
    
    count = 0
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        
        # Save image every 2 seconds
        img_path = os.path.join(FOLDER_NAME, f"image_{count}.jpg")
        cv2.imwrite(img_path, frame)
        print(f"Captured: {img_path}")
        
        count += 1
        time.sleep(2)  # Delay between captures

        # Press 'q' to exit capturing
        if cv2.waitKey(30) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()

# Function to monitor the folder and process images in grayscale
def process_images():
    processed_images = set()
    
    while True:
        images = [f for f in os.listdir(FOLDER_NAME) if f.endswith(".jpg")]
        
        for img in images:
            if img not in processed_images:  # Process only new images
                img_path = os.path.join(FOLDER_NAME, img)
                gray_img_path = os.path.join(FOLDER_NAME, f"gray_{img}")
                
                # Convert to grayscale
                image = cv2.imread(img_path)
                gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
                cv2.imwrite(gray_img_path, gray_image)
                
                print(f"Processed: {gray_img_path}")
                processed_images.add(img)  # Mark as processed

        time.sleep(1)  # Check for new images every second

# Main function to run both processes in parallel
if __name__ == "__main__":
    # Creating two processes
    capture_process = multiprocessing.Process(target=capture_images)
    process_process = multiprocessing.Process(target=process_images)
    
    # Start both processes
    capture_process.start()
    process_process.start()
    
    # Keep them running
    capture_process.join()
    process_process.join()

    print("Both processes completed.")
```
