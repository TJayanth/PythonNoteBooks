```python
# Concept 1 – What is a Class (Basic OOP)

A class is a blueprint for creating objects.
```

```python
class Calculator:

    def add(self, a, b):
        return a + b

    def multiply(self, a, b):
        return a * b
```

```python
calc = Calculator()

print(calc.add(2,3))
print(calc.multiply(4,5))
```

```python
class → blueprint
object → instance of the class
self → refers to the current object
```

```python
# Concept 2 – Writing Functions inside a Class

# Methods are simply functions inside a class.
```

```python
class MathOperations:

    def square(self, x):
        return x * x

    def cube(self, x):
        return x * x * x
```

```python
m = MathOperations()

print(m.square(4))
print(m.cube(3))
```
