```python
# Measures of Central Tendency (Center of the Data)

## These describe the typical value in a dataset.

| Statistic | Meaning                         |
| --------- | ------------------------------- |
| Mean      | Average of all values           |
| Median    | Middle value after sorting      |
| Mode      | Most frequently occurring value |
```

```python
# Measures of Spread (Variability)

## These describe how spread out the data is.

| Statistic                 | Meaning                            |
| ------------------------- | ---------------------------------- |
| Range                     | Max − Min                          |
| Variance                  | Average squared distance from mean |
| Standard Deviation        | Square root of variance            |
| Interquartile Range (IQR) | Spread of the middle 50%           |
```

```python
# Position Statistics
## These describe where values lie in the distribution.

| Statistic  | Meaning                                   |
| ---------- | ----------------------------------------- |
| Percentile | Percentage of data below a value          |
| Quartiles  | Divide data into 4 parts                  |
| Deciles    | Divide data into 10 parts                 |
| Z-score    | Distance from mean in standard deviations |
```

```python
# Shape of Distribution
## These describe the shape of the data

| Statistic | Meaning                   |
| --------- | ------------------------- |
| Skewness  | Asymmetry of distribution |
| Kurtosis  | How heavy the tails are   |
| Outliers  | Values far from the rest  |

```

```python
# Useful Statistics

| Statistic      | Meaning                        |
| -------------- | ------------------------------ |
| Covariance     | Relationship between variables |
| Correlation    | Strength of relationship       |
| Geometric Mean | Useful for growth rates        |
| Harmonic Mean  | Useful for averages of rates   |
```

```python
# Just something about tails 


1. What Are the Tails of a Distribution?

    In a probability distribution:
        Left Tail → very small or unusually low values
        Right Tail → very large or unusually high values
Most data lies in the center, while the tails contain rare events.
```

```python
        Tail        Center        Tail
         |            |            |
----|----|----|----|----|----|----|----
   very low      normal        very high
```

```python
Why the Tails Are Important  ?

---- > Rare Events Occur in the Tails
Unusual or extreme events live here.
Examples:
•	Financial market crashes
•	System failures
•	Medical anomalies
•	Fraud detection
•	Extreme weather
The mean and median describe normal behaviour, but tails describe unusual behaviour.
 
-----> Most Risks Come from the Tails
In many systems, extreme outcomes cause the largest damage.
Examples:
| Domain        | Tail Event           |
| ------------- | -------------------- |
| Finance       | market crash         |
| Engineering   | catastrophic failure |
| Medicine      | rare disease         |
| Cybersecurity | attack or breach     |
| AI            | model hallucination  |

These events happen rarely, but when they occur they have huge consequences.
 
------> Innovation Often Starts in the Tails

The tails contain outliers, and outliers can represent:
•	mistakes
•	anomalies
•	new discoveries

Examples:
| Example              | Tail Observation             |
| -------------------- | ---------------------------- |
| Penicillin discovery | unexpected mold growth       |
| Physics discoveries  | unexpected experimental data |
| Fraad detection      | unusual transaction patterns |
| Customer insights    | rare but high-value users    |

Sometimes the most valuable insights come from studying anomalies.
 
------> Tails Reveal Distribution Shape

The tails tell us about skewness and risk.
Example distributions:
| Distribution | Tail Behavior         |
| ------------ | --------------------- |
| Normal       | thin tails            |
| Heavy-tailed | more extreme events   |
| Power-law    | extreme events common |

Example:
    Human height → thin tails
    Wealth distribution → heavy right tail

A few people hold extreme wealth.
 
------> Machine Learning and AI Depend on Tail Understanding

Models trained on average cases can fail on edge cases.

Examples:
•	self-driving cars
•	medical diagnosis
•	fraud detection

The difficult cases often lie in the tails.
This is why AI researchers focus on edge-case testing.

```

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(0,1,10000)

plt.hist(data, bins=50)
plt.title("Normal Distribution with Tails")
plt.show()
```
